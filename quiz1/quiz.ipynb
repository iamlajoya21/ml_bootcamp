```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
student_data = [
    ("Karen Poghosyan", 0),
    ("Arsen Sirunyan", 83),
    ("Tigran Hakobyan", 0),
    ("Eliza Asatryan", 63),
    ("Yulya Sargsyan", 40),
    (" ", 48),
    ("Mels Tadevosyan", 75),
    ("Areg Ayvazyan", 83),
    ("Astghik Saghyan", 60),
    ("Ashot Klekchyan", 48),
    ("Lilith", 23),
    (" ", 8),
    ("Astghik Saghyan(1)", 50),
    ("Samvel Harutunyan", 43),
    ("Armen Gasparyan", 16)
]

# Extract names and grades
students = [data[0] for data in student_data]
grades = [data[1] for data in student_data]

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Bar Chart (sorted by grade)
sorted_indices = np.argsort(grades)[::-1]  # Sort descending
sorted_students = [students[i] for i in sorted_indices]
sorted_grades = [grades[i] for i in sorted_indices]

bars = ax1.bar(range(len(sorted_students)), sorted_grades, color='skyblue', edgecolor='black')
ax1.set_xlabel('Students', fontsize=12)
ax1.set_ylabel('Grade (out of 100)', fontsize=12)
ax1.set_title('Student Grades Bar Chart', fontsize=14, fontweight='bold')
ax1.set_xticks(range(len(sorted_students)))
ax1.set_xticklabels(sorted_students, rotation=90, fontsize=8)
ax1.set_ylim(0, 100)
ax1.grid(axis='y', alpha=0.3)

# Add grade values on top of bars
for bar, grade in zip(bars, sorted_grades):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{grade}', ha='center', va='bottom', fontsize=8)

# Histogram
ax2.hist(grades, bins=10, edgecolor='black', color='lightcoral', alpha=0.7)
ax2.set_xlabel('Grade (out of 100)', fontsize=12)
ax2.set_ylabel('Frequency', fontsize=12)
ax2.set_title('Grade Distribution Histogram', fontsize=14, fontweight='bold')
ax2.set_xlim(0, 100)
ax2.grid(alpha=0.3)

# Add statistics text
mean_grade = np.mean(grades)
median_grade = np.median(grades)
stats_text = f'Mean: {mean_grade:.1f}\nMedian: {median_grade:.0f}'
ax2.text(0.95, 0.95, stats_text, transform=ax2.transAxes,
         fontsize=10, verticalalignment='top',
         horizontalalignment='right',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()

# Print some statistics
print("Grade Statistics:")
print(f"Total students: {len(grades)}")
print(f"Average grade: {mean_grade:.1f}")
print(f"Median grade: {median_grade}")
print(f"Highest grade: {max(grades)}")
print(f"Lowest grade: {min(grades)}")
print(f"Number of failing grades (<60): {sum(1 for g in grades if g < 60)}")
print(f"Pass rate (% with 60): {(sum(1 for g in grades if g >= 60)/len(grades)*100):.1f}%")
```
