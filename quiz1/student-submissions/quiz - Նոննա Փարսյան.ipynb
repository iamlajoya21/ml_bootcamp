<h6>Problem 2: The Login Security Monitor (15 Minutes)
Topics: Sets, Tuples, List Filtering, File I/O.

The Task: You have a log of 12 login attempts. Some are from blacklisted IPs.

Filter: Remove any IP address present in the blacklisted_ips set.

Deduplicate: Ensure each IP appears only once.

Store: Save the safe, unique IPs into a Tuple.

Write: Write the safe IPs to authorized_logins.txt, one per line.
</h6>


```python
all_logins = [
    "192.168.1.1", "10.0.0.5", "192.168.1.1", "172.16.0.1", 
    "10.0.0.5", "192.168.1.10", "172.16.0.1", "8.8.8.8", 
    "192.168.1.1", "10.0.0.9", "192.168.1.10", "10.0.0.5"
]
blacklisted_ips = {"172.16.0.1", "8.8.8.8", "200.50.10.1"}
unique_ips=set()
for item in all_logins:
    if item in blacklisted_ips:
        del item
    else:
        unique_ips.add(item)
safe_ips=tuple(unique_ips)
print(safe_ips)
for item in safe_ips:
    with open('authorized_logins.txt','w') as authorized_logins:
        authorized_logins.write(item +'\n')
```

<h6>Problem 3: The "Strict" Transaction Processor (15 Minutes)



The Task: Create a terminal-based transaction logger.

Loop: Use while True to ask for an amount. Exit only if the user types "exit".

Validation: * If the input isn't a number, catch ValueError and print "Invalid Data Type".

Manual Raise: If the number is greater than 5000, raise a ValueError with the message "High Value: Requires Approval".

Manual Raise: If the number is 0 or less, raise a ValueError with the message "Invalid Amount".

Finally: After every attempt (valid or invalid), print "System: Monitoring active."

</h6>



```python
try:
    while True:
        amount=input("Enter amount:")
        if amount=='exit':
            break 
        try:
            if int(amount)>5000:
                raise ValueError()
        except ValueError:
            print("High Value: Requires Approval")
        try:  
            if int(amount)<=0:
                raise ValueError()
        except ValueError:
            print("Invalid Amount")
except ValueError:   
    print("Invalid Data Type")
finally:
    print("System:Monitoring active")
        
```

<h6>Problem 4: The Global Gradebook Merger (10 Minutes) 


The Task: You are merging records from two different semesters.

Logic: Create a single dictionary where the Key is the student name and the Value is a List of all grades they earned across both semesters.

Requirement: Do not overwrite data. If "Alice" is in both, her new list should contain all 4 grades.</h6>

```python
semester_1 = {
    "Alice": [85, 90], "Bob": [70, 65], "Charlie": [100], 
    "David": [88], "Eve": [92, 91]
}
semester_2 = {
    "Alice": [95, 88], "Frank": [78, 80], "Bob": [72], 
    "Grace": [99], "Charlie": [90, 85], "Heidi": [75]
}
all_grades={}
for name,score in semester_1.items():
    all_grades[name]=score
for name,score in semester_2.items():
    if name not in all_grades:
        all_grades[name]=score
    else:
        all_grades[name].extend(score)
print(all_grades)
```
