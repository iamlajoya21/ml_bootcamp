```python
#problem 1
final_result ={}
repeat = 0
try:
    with open('inventory.txt', 'r') as invent, open('audit_errors.log', 'w') as audit:
        qunatity = 0
        for line in invent:
            parts = line.strip().split(',')
            try:
                if len(parts)==2:
                    item = parts[0].uppercase()
                    num = int(parts[1])
                    final_result[item]=num
            except ValueError:
                audit.write(line +"\n")
                print(f"the errors {audit}")
except FileNotFoundError:
    print("Could not find raw_data.txt")
finally:
    print(final_result)
               
```

```python
#problem 2


all_logins = [
    "192.168.1.1", "10.0.0.5", "192.168.1.1", "172.16.0.1", 
    "10.0.0.5", "192.168.1.10", "172.16.0.1", "8.8.8.8", 
    "192.168.1.1", "10.0.0.9", "192.168.1.10", "10.0.0.5"
]
blacklisted_ips = {"172.16.0.1", "8.8.8.8", "200.50.10.1"}

final = set()
for login in all_logins:
    if login not in blacklisted_ips:
        final.add(login)
print(final)

with open('authorized_logins.txt','w') as f:
    for line in final:
        f.write(line + '\n')
```

```python
#problem 4
semester_1 = {
    "Alice": [85, 90], "Bob": [70, 65], "Charlie": [100], 
    "David": [88], "Eve": [92, 91]
}
semester_2 = {
    "Alice": [95, 88], "Frank": [78, 80], "Bob": [72], 
    "Grace": [99], "Charlie": [90, 85], "Heidi": [75]
}

final = {}
names = list(set(list(semester_1.keys()) + list(semester_2.keys())))
for name in names:
    grades = []
    
    if name in semester_1:
        grades.extend(semester_1[name])
    if name in semester_2:
        grades.extend(semester_2[name])
    
    final[name] = grades

print(final)
```
