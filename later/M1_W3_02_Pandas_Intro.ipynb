## 1. Under the Hood of a DataFrame

### Columnar Storage
Each column is a NumPy array. Efficient for column-wise math.

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

# Load Real Data: California Housing
housing = fetch_california_housing(as_frame=True)
df = housing.frame

print(df.head())
```

## 2. Selection: loc vs iloc

Label vs Position.

```python
# Select rows 0 to 5 and columns 'MedInc' and 'HouseAge'
print(df.loc[0:5, ['MedInc', 'HouseAge']])
```

## 3. The Query Method

Filtering made easy.

```python
# Houses with more than 5 bedrooms (Avg) and price > 4.0 (400k)
luxury = df.query("AveRooms > 5 and MedHouseVal > 4.0")
print(f"Found {len(luxury)} luxury districts.")
```
