## 1. Operating System Interaction: Files

### File Descriptors
When you open a file, Python asks the OS kernel to open the file. The OS returns an integer handle (File Descriptor). Python wraps this integer in a file object.

### Text vs Binary Mode
- **Binary (`rb`)**: Returns raw bytes (`b'\xe4'`). This is exactly what is on the disk.
- **Text (`r`)**: Python decodes the bytes into a string using an encoding (usually UTF-8).

### Buffering
Disk I/O is slow. Writing to disk for every single character would be terribly inefficient. 
- **Buffered I/O**: Python writes to a memory buffer (e.g., 4KB or 8KB). When the buffer is full, or when `flush()`/`close()` is called, the chunk is written to disk in one go.

## 2. Modern File Handling (pathlib)

`pathlib` treats paths as objects, abstracting away OS differences (Windows `\` vs Linux `/`).

```python
from pathlib import Path

# Current directory
p = Path('.')

# Check if file exists
f = Path("example.txt")
if not f.exists():
    f.write_text("Hello Pathlib!", encoding="utf-8")

print(f.read_text(encoding="utf-8"))
```

## 3. Context Managers (The `with` statement)

The `with` statement guarantees that resources are released (files closed) even if an error occurs. It calls `__enter__` at the start and `__exit__` at the end.

```python
with open("example.txt", "r") as f:
    # f.__enter__() called
    content = f.read()
    # f.__exit__() called automatically here
```

## 4. Serialization (JSON)

JSON is text-based. It maps almost 1:1 with Python dictionaries and lists, making it the standard for web APIs.

```python
import json

data = {"name": "Alice", "skills": ["Python"]}
json_str = json.dumps(data)
print(type(json_str), json_str)
```
