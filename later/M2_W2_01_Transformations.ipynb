## 1. Log-Normality in Space

We'll use the **Planets** dataset (Exoplanets discovered). The `orbital_period` often spans orders of magnitude.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

planets = sns.load_dataset('planets')
data = planets['orbital_period'].dropna()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.hist(data, bins=30)
ax1.set_title("Original (Skewed)")

ax2.hist(np.log1p(data), bins=30)
ax2.set_title("Log-Transformed (Normal-ish)")
plt.show()
```
