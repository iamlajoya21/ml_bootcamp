## 1. The Anatomy of a Function Call

### The Call Stack
When a function is called, a "Stack Frame" is pushed onto the Call Stack. This frame contains:
- Local variables.
- Arguments passed.
- The return address (where to go back to).
When the function returns, the frame is popped (destroyed). This is why local variables disappear.

### First-Class Citizens
In Python, functions are objects just like integers or strings. You can:
- Assign them to variables.
- Pass them as arguments (Higher-Order Functions).
- Return them from other functions (Closures/Decorators).

## 2. Advanced Arguments

### *args (Tuple packing)
Allows variable positional arguments.

```python
def sum_all(*args):
    # args is a tuple inside the function
    return sum(args)

print(sum_all(1, 2, 3))
```

### **kwargs (Dictionary packing)
Allows variable keyword arguments.

```python
def print_profile(**kwargs):
    # kwargs is a dict inside the function
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_profile(name="Alice", age=25)
```

## 3. Scope and LEGB Rule

Python resolves variable names in this order:
1. **L**ocal (Inside function)
2. **E**nclosing (Inside outer function, relevant for closures)
3. **G**lobal (Module level)
4. **B**uilt-in (Python keywords like `len`, `print`)

```python
x = "Global"

def outer():
    x = "Enclosing"
    def inner():
        x = "Local"
        print(x)
    inner()
    
outer()
```

## 4. Decorators Under the Hood

A decorator is syntactic sugar for `func = decorator(func)`. It uses the concept of closures (functions returning functions that remember their scope).

```python
def my_decorator(func):
    def wrapper():
        print("Before call")
        func()
        print("After call")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```
