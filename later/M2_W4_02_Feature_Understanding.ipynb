## 1. Feature Engineering on Titanic

Real scenarios: Scaling Fare, Encoding Embarked.


```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder

df = sns.load_dataset('titanic')[['fare', 'age', 'sex', 'embarked']].dropna()

# Scale Numerical
scaler = StandardScaler()
df[['fare', 'age']] = scaler.fit_transform(df[['fare', 'age']])

# Encode Categorical
print("Before OneHot:\n", df.head())
enc = OneHotEncoder(sparse=False)
encoded_cats = enc.fit_transform(df[['sex', 'embarked']])
print("Encoded Shape:", encoded_cats.shape)
```
