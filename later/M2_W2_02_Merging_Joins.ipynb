## 1. Splitting a Dataset to Join it Back

We will split the `tips` dataset to simulate a relational database.
1. **Orders Table**: total_bill, tip, size, time, day.
2. **Customers Table**: sex, smoker.

```python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')
tips['order_id'] = range(len(tips))

orders = tips[['order_id', 'total_bill', 'tip']]
customers = tips[['order_id', 'sex', 'smoker']]

# Merge them back
merged = pd.merge(orders, customers, on='order_id')
print(merged.head())
```

## 2. Set Theory

Demonstrating anti-joins.

```python
# Simulate missing customers
incomplete_customers = customers.iloc[:200]  # Drop last 44 customers

# Find orders without customer info (Anti-Join)
lost = pd.merge(orders, incomplete_customers, on='order_id', how='left', indicator=True)
missing_info = lost[lost['_merge'] == 'left_only']
print(f"Orders with missing customer info: {len(missing_info)}")
```
