## 1. Advanced Conditions & Boolean Logic

### Short-Circuit Evaluation
Python evaluates logical expressions from left to right and stops as soon as the result is determined.
- `False and X`: Returns `False` without looking at X.
- `True or X`: Returns `True` without looking at X.

This is why you can safely write `if obj is not None and obj.value > 5`. If `obj` is None, the second part (which would crash) is never executed.

### Truthy and Falsy Table
| Data Type | Falsy Value |
|-----------|-------------|
| int       | 0           |
| float     | 0.0         |
| str       | ""          |
| list      | []          |
| dict      | {}          |
| NoneType  | None        |

```python
def expensive_check():
    print("I am running!")
    return True

print("Test 1:")
if True or expensive_check():
    print("Done")
    
print("\nTest 2:")
if False and expensive_check():
    print("Done")
```

### Match-Case (Structural Pattern Matching)
Use `match-case` when checking a value against multiple patterns. It is more optimized than long `if-elif-else` chains for this purpose.

```python
status = 404

match status:
    case 200:
        print("OK")
    case 404:
        print("Not Found")
    case _:
        print("Unknown")
```

## 2. Advanced Loops & Complexity

### The Danger of Nested Loops (Big O)
When you put a loop inside a loop, your complexity often becomes $O(n^2)$. 
- One loop over 1000 items = 1,000 operations.
- Nested loop over 1000 items = 1,000,000 operations.

Always prefer flat structures or linear comprehensions when possible.

```python
# O(n^2) Quadratic Time
count = 0
for i in range(100):
    for j in range(100):
        count += 1
print(count)
```

### List Comprehensions: Under the Hood
Comprehensions are not just syntactic sugar; they are often faster than for-loops because the loop overhead is handled in C rather than Python bytecode interpretation step-by-step.

```python
# Comprehension
squares_comp = [x**2 for x in range(10)]
print(squares_comp)
```
