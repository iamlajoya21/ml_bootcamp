## 1. The "Split-Apply-Combine" Strategy

### Theory
Splitting data into groups, applying a function, and combining results.

```python
import pandas as pd
import seaborn as sns
import numpy as np

# Load Real Data: Tips dataset
df = sns.load_dataset('tips')
print(df.head())
```

## 2. Pivot Tables

Analysing tips by Day and Sex.

```python
pivot = df.pivot_table(index="day", columns="sex", values="total_bill", aggfunc="mean")
print("Average Total Bill by Day and Sex:\n", pivot)
```

## 3. Window Functions

To use rolling functions, we need a time index. Let's assume these tips came in sequentially.

```python
df['rolling_avg_bill'] = df['total_bill'].rolling(window=10).mean()
print(df[['total_bill', 'rolling_avg_bill']].head(15))
```
