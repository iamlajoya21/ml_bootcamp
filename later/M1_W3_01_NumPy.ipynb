## 1. Why is NumPy so Fast?

### Contiguous Memory vs Linked Lists
- **Python Lists**: An array of pointers to objects scattered in memory. Iterating means jumping around memory (cache misses).
- **NumPy Arrays**: A solid block of contiguous memory. The items are stored next to each other.

### SIMD
Single Instruction, Multiple Data. CPU vector operations.

```python
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load Real Data: The Digits dataset (8x8 images of handwritten numbers)
digits = load_digits()
data = digits.images

print(f"Data Shape: {data.shape}")
print("First Image (8x8 array):\n", data[0])

plt.imshow(data[0], cmap='gray')
plt.title(f"Target: {digits.target[0]}")
plt.show()
```

## 2. Broadcasting Rules

Align dimensions by padding 1s.

```python
# Normalize pixel intensity (0-16) to (0-1) using broadcasting
# We divide the entire (1797, 8, 8) array by a scalar (16.0)
normalized_data = data / 16.0
print(f"Max value after norms: {normalized_data.max()}")
```

## 3. Boolean Masking

Filtering data.

```python
# Select only images that represent the number '3'
threes = data[digits.target == 3]
print(f"Found {len(threes)} images of the number 3")

plt.imshow(threes[0], cmap='gray')
plt.title("The first '3'")
plt.show()
```
