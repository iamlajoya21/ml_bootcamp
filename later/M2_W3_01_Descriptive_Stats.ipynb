## 1. Skewness and Kurtosis (Tips Dataset)

We use the **Tips** dataset. 'Tip' amounts are naturally right-skewed (cannot be negative, most are small, some are huge).

```python
import pandas as pd
import seaborn as sns
from scipy import stats

tips = sns.load_dataset('tips')

print("Skewness of Tip:", tips['tip'].skew())
print("Kurtosis of Tip:", tips['tip'].kurt())
```

## 2. Robust Statistics

Comparing Mean vs Trimmed Mean on Tips.

```python
print("Mean Tip:", tips['tip'].mean())
print("Trimmed Mean (10%):", stats.trim_mean(tips['tip'], 0.1))
```

## 3. Correlation

Total Bill vs Tip.


```python
print(tips[['total_bill', 'tip']].corr(method='pearson'))
```
