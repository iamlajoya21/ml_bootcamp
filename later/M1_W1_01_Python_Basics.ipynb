## 1. Syntax Overview & Memory Model: Under the Hood

### The CPython Interpreter
When you write Python code, you are technically writing instructions for a virtual machine represented by the CPython interpreter (written in C). It manages memory, parses code, and executes bytecode.

### Everything is an Object (PyObject)
In languages like C or Java, `int x = 5` reserves a 32-bit memory space for the integer 5. In Python, `x = 5` is much more complex. `5` is an instance of a `PyObject` struct in C, containing:
- `ob_refcnt`: Reference count (for garbage collection).
- `ob_type`: Pointer to the type type object (int).
- `ob_digit`: The actual value.

**Key Concept**: Variable names in Python are just **labels** (pointers) to these objects in the heap memory. They are not boxes containing data.

### Reference Counting
Python manages memory automatically via Reference Counting. Every time you assign a variable to an object, its reference count goes up. When it goes to zero, the memory is freed.

````python
import sys

# Create an object
a = []
b = a  # Reference count increases

print(f"Reference count for list: {sys.getrefcount(a)}") 
# Note: getrefcount returns +1 because passing the arg creates a temp reference
````

## 2. Variables, Types & Mutability (Advanced)

### Mutability and Memory
- **Immutable Objects**: When you "change" an int, you are actually creating a *new* int object and moving the label.
- **Mutable Objects**: Lists/Dicts act as containers of pointers. Changing an element changes which object a slot points to, but the container's memory address remains the same.

### Why is Python "Strongly Typed"?
Even though you don't declare types (Dynamic), Python enforces types at runtime. You cannot add a string `"5"` to an integer `5` without explicit casting. This prevents implicit type coercion errors found in JavaScript.

````python
# Immutable example
x = 10
old_id = id(x)
x = x + 1
new_id = id(x)
print(f"Immutable Int changed ID: {old_id != new_id}")

# Mutable example
lst = [1]
old_id = id(lst)
lst.append(2)
new_id = id(lst)
print(f"Mutable List kept ID: {old_id == new_id}")
````

## 3. Advanced String Formatting (f-strings)

### How f-strings work
F-strings are evaluated at runtime, not constant time. This means you can put valid Python expressions inside `{}`. They are faster than `%` formatting or `.format()` because they are generated as part of the bytecode instructions directly.

````python
name = "Alice"
score = 95.2356

# Basic
print(f"Hello {name}!")

# Expressions inside brackets
print(f"Next year {name} will be {20 + 5} years old.")

# Formatting numbers (2 decimal places)
print(f"Score: {score:.2f}")

# Alignment (Right align 10 chars)
print(f"|{name:>10}|")
````

## 4. Introduction to Type Hinting

### Static Analysis vs Runtime
Type hints (`age: int`) do **not** affect runtime behavior. CPython ignores them. However, they allow static analysis tools (like MyPy) to check your code for errors *before* you run it, bringing safety to large codebases.

````python
name: str = "Bob"
age: int = 30
height: float = 1.75
is_student: bool = True

# Even complex types
from typing import List, Dict

scores: List[int] = [10, 20, 30]
user_info: Dict[str, str] = {"name": "Alice"}
````
