## 1. The Theory of Missing Data (Titanic Case Study)

We will use the Titanic dataset. 
- **Age**: Standard missingness (MAR? MCAR?).
- **Deck**: Heavily missing (MNAR?).

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
print(df.isnull().sum())
```

## 2. Imputation Strategies

Imputing Age based on Class (Pclass).

```python
# Age distribution before
print("Mean Age before:", df['age'].mean())

# Impute using transform within Pclass
df['age'] = df.groupby('pclass')['age'].transform(lambda x: x.fillna(x.median()))

print("Mean Age after:", df['age'].mean())
```
