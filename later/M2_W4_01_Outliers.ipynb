## 1. Multivariate Outliers (Housing)

Use California Housing. High AveRooms but low Price? Outlier.


```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

housing = fetch_california_housing(as_frame=True)
df = housing.frame
X = df[['AveRooms', 'MedHouseVal']]

clf = IsolationForest(contamination=0.01, random_state=42)
preds = clf.fit_predict(X)

plt.scatter(X['AveRooms'], X['MedHouseVal'], c=preds, cmap='coolwarm', alpha=0.6)
plt.xlabel("Ave Rooms")
plt.ylabel("Price")
plt.title("Housing Outliers")
plt.xlim(0, 20) # Focusing on readable range
plt.show()
```
