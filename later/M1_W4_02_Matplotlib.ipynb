## 1. The Hierarchy of Objects

Figure -> Axes -> Axis -> Artist.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load Real Data: Iris
iris = sns.load_dataset('iris')
print(iris.head())
```

## 2. Subplots with Real Data

Let's compare Petal and Sepal dimensions.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Scatter: Sepal Length vs Width
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    ax1.scatter(subset['sepal_length'], subset['sepal_width'], label=species)
ax1.set_title("Sepal Dimensions")
ax1.set_xlabel("Length")
ax1.set_ylabel("Width")
ax1.legend()

# Hist: Petal Length Distribution
ax2.hist(iris['petal_length'], bins=20, color='purple', alpha=0.7)
ax2.set_title("Petal Length Distribution")

plt.show()
```
