Հայերեն: Մենք պետք է իմանանք, թե Նյու Յորքի որ թաղամասերն ունեն ամենաշատ հայտարարությունները: Խմբավորեք տվյալները ըստ թաղամասերի խմբի (neighbourhood_group) և հաշվեք յուրաքանչյուրում առկա հայտարարությունների ընդհանուր քանակը:

Problem 2: Room Type Economics / Սենյակների տեսակների տնտեսագիտությունը
English: Find the Average Price for each room_type. Sort the result from most expensive to cheapest.
Հայերեն: Գտեք Միջին գինը յուրաքանչյուր սենյակի տեսակի (room_type) համար: Դասավորեք արդյունքները ամենաթանկից մինչև ամենաէժանը:

Problem 3: The "Super-Hosts" / «Սուպեր-տանտերերը»
English: Some hosts manage multiple properties. Group by host_id and host_name, calculate the total listings for each, and show only the Top 5 hosts.
Հայերեն: Որոշ տանտերեր կառավարում են բազմաթիվ բնակարաններ: Խմբավորեք ըստ տանտիրոջ ID-ի և անվան, հաշվարկեք յուրաքանչյուրի հայտարարությունների քանակը և ցուցադրեք միայն լավագույն 5 տանտերերին:

Problem 4: Neighborhood Price Spans / Գների միջակայքը թաղամասերում
English: For each neighbourhood_group, find the Maximum, Minimum, and Median price in a single summary table.
Հայերեն: Յուրաքանչյուր թաղամասի խմբի համար գտեք Առավելագույն, Նվազագույն և Մեդիան գները մեկ ամփոփիչ աղյուսակում:

Problem 5: Popularity Contest / Հանրաճանաչության մրցույթ
English: Calculate the Total Number of Reviews for each neighbourhood_group. Which group is the most "talked about"?
Հայերեն: Հաշվարկեք հետադարձ կապի (number_of_reviews) ընդհանուր քանակը յուրաքանչյուր թաղամասի խմբի համար: Ո՞ր խումբն է ամենաշատ «քննարկվում»:

Problem 6: Manhattan Availability Audit / Մանհեթենի հասանելիության աուդիտ
English: Focus only on Manhattan. Find the Average Availability (availability_365) for each specific neighbourhood within Manhattan.
Հայերեն: Կենտրոնացեք միայն Մանհեթենի վրա: Գտեք Միջին հասանելիությունը (availability_365) Մանհեթենի յուրաքանչյուր կոնկրետ թաղամասի (neighbourhood) համար:

Problem 7: Multi-Level Pricing Strategy / Գնագոյացման բազմամակարդակ ռազմավարություն
English: Group the data by both neighbourhood_group and room_type. Calculate the Average Price and the Total Number of Reviews for every combination.
Հայերեն: Խմբավորեք տվյալները թե՛ թաղամասերի խմբի, թե՛ սենյակի տեսակի հիման վրա: Հաշվարկեք Միջին գինը և Հետադարձ կապի ընդհանուր քանակը յուրաքանչյուր համակցության համար:

Problem 8: The Value Seekers / Արժեք փնտրողները
English: For each neighbourhood_group, calculate the Average Price and the Average Number of Reviews. Then, identify the group that has the highest reviews but the lowest average price (the best "value" area).
Հայերեն: Յուրաքանչյուր թաղամասի խմբի համար հաշվարկեք Միջին գինը և Հետադարձ կապի միջին քանակը: Այնուհետև պարզեք, թե որ խումբն ունի ամենաշատ կարծիքները, բայց ամենացածր միջին գինը (ամենաշահավետ տարածքը):

Problem 9: High-Traffic Neighborhoods / Մեծ հոսք ունեցող թաղամասերը
English: Find the Average Price for every neighbourhood, but only include neighborhoods that have more than 100 listings. Sort the result by price.
Հայերեն: Գտեք Միջին գինը յուրաքանչյուր թաղամասի համար, բայց ներառեք միայն այն թաղամասերը, որոնք ունեն ավելի քան 100 հայտարարություն: Դասավորեք արդյունքն ըստ գնի:

Problem 10: The Business Host Deep-Dive / Պրոֆեսիոնալ տանտերերի վերլուծություն
English: Identify hosts who have more than 10 listings (calculated by calculated_host_listings_count). For this specific group of "Business Hosts," find their Average Price per night and the Total Revenue they generate (Price * Number of Reviews as a proxy for revenue).

```python
import pandas as pd
import numpy as np
import ssl

# Setup
ssl._create_default_https_context = ssl._create_unverified_context
url = "https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/airbnb.csv"
df = pd.read_csv(url)

# --- PRE-PROCESSING (Problem 2 Logic) ---
# Cleaning price more robustly to ensure it's numeric
df['price'] = df['price'].str.replace(r'\D', '', regex=True).astype(float)

#=============================================================================
# Problem 1: Neighbourhood Listings / Բնակարանների քանակը
# Optimization: Named aggregation is great; .head() shows the result.
neighbourhood_listing = df.groupby('neighbourhood_full').agg(total_listings=('listing_id', 'count'))

#=============================================================================
# Problem 2: Average Price / Միջին գինը
# Optimization: Sorted and stored for a clean view.
avg_price_sorted = df.groupby('room_type')['price'].mean().sort_values(ascending=False)

#=============================================================================
# Problem 3: Top Hosts / Լավագույն տանտերերը
# Optimization: Using .size() is faster than .count() when you just need row counts.
top_hosts = df.groupby(['host_id', 'host_name']).size().reset_index(name='count').sort_values('count', ascending=False)

#=============================================================================
# Problem 4: Price Summary / Գների ամփոփագիր
# Optimization: Using strings for 'max', 'min' and np.median for accuracy.
summary = df.groupby('neighbourhood_full')['price'].agg(['max', 'min', np.median])

#=============================================================================
# Problem 5: Most Talked About / Ամենաշատ քննարկվողը
# Optimization: 'Most talked about' usually refers to the total SUM of reviews.
df_most_talked = df.groupby('neighbourhood_full')['number_of_reviews'].sum().sort_values(ascending=False).head(1)

#=============================================================================
# Problem 6: Manhattan Availability / Մանհեթենի հասանելիությունը
# Optimization: Case-insensitive search for Manhattan.
manhattan_df = df[df['neighbourhood_full'].str.contains('Manhattan', case=False, na=False)]
avg_avail_man = manhattan_df.groupby('neighbourhood_full')['availability_365'].mean()

#=============================================================================
# Problem 7: Reviews & Price / Կարծիքներ և գին
df_7 = df.groupby('neighbourhood_full').agg(avg_price=('price', 'mean'), total_reviews=('number_of_reviews', 'sum'))

#=============================================================================
# Problem 8: The Best Value / Ամենաշահավետ տարբերակը
# Optimization: High reviews + Low price = Value.
df_8 = df.groupby('neighbourhood_full').agg(avg_price=('price', 'mean'), avg_reviews=('number_of_reviews', 'mean'))
df_8 = df_8.sort_values(by=['avg_price', 'avg_reviews'], ascending=[True, False])

#=============================================================================
# Problem 9: Popular Neighborhoods (>100) / Հանրահայտ թաղամասեր
# Optimization: Avoid .filter() for speed; use boolean masking on the aggregated DF.
df_9 = df.groupby('neighbourhood_full').agg(total_listings=('listing_id', 'count'), avg_price=('price', 'mean'))
df_9 = df_9[df_9['total_listings'] > 100]

#=============================================================================
# Problem 10: Professional Host Revenue / Պրոֆեսիոնալ տանտերերի եկամուտը
# Optimization: Streamlined revenue calculation and filtering.
df['revenue'] = df['price'] * df['number_of_reviews']
df_10 = df.groupby(['host_id', 'host_name']).agg(
    avg_price=('price', 'mean'),
    total_revenue=('revenue', 'sum'),
    total_listings=('listing_id', 'count')
)
df_10 = df_10[df_10['total_listings'] > 10]
df_10
```
