```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import mean_squared_error, accuracy_score

# --- SOLUTION: CHALLENGE 1 (DIAMONDS) ---
diamonds = sns.load_dataset('diamonds')
# Cleaning
diamonds = diamonds[(diamonds[['x','y','z']] != 0).all(axis=1)]
# Engineering
diamonds['volume'] = diamonds['x'] * diamonds['y'] * diamonds['z']
# Scaling
scaler = MinMaxScaler()
num_cols = ['carat', 'depth', 'table', 'volume']
diamonds[num_cols] = scaler.fit_transform(diamonds[num_cols])

X1 = diamonds[num_cols]
y1 = diamonds['price']

# Grid Search for Linear Regression
param_grid_lr = {'fit_intercept': [True, False]}
grid_lr = GridSearchCV(LinearRegression(), param_grid_lr, cv=5, scoring='neg_mean_squared_error')
grid_lr.fit(X1, y1)
print(f"Diamond MSE (CV): {-grid_lr.best_score_:.2f}")

# --- SOLUTION: CHALLENGE 2 (TITANIC) ---
titanic = sns.load_dataset('titanic')
# Cleaning & Imputation
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
titanic.drop(['deck', 'embark_town', 'alive', 'who', 'adult_male'], axis=1, inplace=True)
titanic.dropna(inplace=True)
# Engineering
titanic['family_size'] = titanic['sibsp'] + titanic['parch']
# Encoding
le = LabelEncoder()
titanic['sex'] = le.fit_transform(titanic['sex'])
titanic['class'] = le.fit_transform(titanic['class'])

X2 = titanic[['pclass', 'sex', 'age', 'family_size', 'fare']]
y2 = titanic['survived']
X2_scaled = scaler.fit_transform(X2)

# Logistic Regression Tuning
param_grid_log = {'C': [0.1, 1, 10], 'penalty': ['l2']}
grid_log = GridSearchCV(LogisticRegression(), param_grid_log, cv=5)
grid_log.fit(X2_scaled, y2)

# Naive Bayes CV
nb_scores = cross_val_score(GaussianNB(), X2_scaled, y2, cv=5)

print(f"Logistic Regression Accuracy: {grid_log.best_score_:.4f}")
print(f"Naive Bayes Accuracy: {nb_scores.mean():.4f}")
```
