## Problem 1: The Volatile Stock Market / Անկայուն Ֆոնդային Շուկա
English: You have an array of 60 days of stock prices. However, on some days, the stock exchange was closed or the data was not recorded, resulting in NaN values.

Create an array of 60 values using np.random.uniform(150, 200, 60).

Randomly replace 10 of these values with np.nan.

Calculate the average stock price ignoring the NaN values.

Replace all NaN values with the average price you calculated in step 3.

Use a ufunc to calculate the daily price change (log return) using np.log().

### Հայերեն: 
Դուք ունեք բաժնետոմսերի գների 60 օրվա տվյալներ: Սակայն որոշ օրերին բորսան փակ է եղել կամ տվյալները չեն գրանցվել, ինչի հետևանքով առաջացել են NaN արժեքներ:

Ստեղծեք 60 արժեքից բաղկացած զանգված՝ օգտագործելով np.random.uniform(150, 200, 60):

Այս արժեքներից 10-ը պատահականորեն փոխարինեք np.nan-ով:

Հաշվեք բաժնետոմսերի միջին գինը՝ անտեսելով NaN արժեքները:

Բոլոր NaN արժեքները փոխարինեք 3-րդ կետում հաշվարկված միջին գնով:

Օգտագործեք ufunc՝ հաշվարկելու օրական գնի փոփոխությունը (լոգարիթմական եկամտաբերություն)՝ կիրառելով np.log():

```python
import numpy as np

# 1. Create 60 days of prices / Ստեղծել 60 օրվա գներ
np.random.seed(42)
prices = np.random.uniform(150, 200, 60)

# 2. Randomly insert 10 NaNs / Պատահականորեն տեղադրել 10 NaN
nan_indices = np.random.choice(60, 10, replace=False)
prices[nan_indices] = np.nan

# 3. Calculate mean ignoring NaNs / Հաշվել միջինը՝ անտեսելով NaN-երը
# We use np.nanmean because np.mean would return NaN if any NaN is present
avg_price = np.nanmean(prices)

# 4. Impute (fill) NaNs with the average / Լրացնել NaN-երը միջին արժեքով
# np.isnan(prices) creates a boolean mask where True = missing data
prices[np.isnan(prices)] = avg_price

# 5. Daily log return / Օրական լոգարիթմական եկամտաբերություն
# np.log() is a ufunc. log(price) helps analyze percentage changes
log_returns = np.log(prices)

print(f"Average Price (Mean): {avg_price:.2f}")
print(f"Log Returns (first 5): {log_returns[:5]}")
```

## Problem 2: Forest Fire Prevention / Անտառային Հրդեհների Կանխարգելում
English: A forest has 100 humidity sensors. Low humidity indicates a high risk of fire.

Generate an array of 100 humidity percentages (0% to 100%) using np.random.randint.

Use comparison operators to create a boolean mask for "Danger Zones" where humidity is below 20%.

Count how many sensors are in Danger Zones.

Use np.where to create a new array called action_plan, where any sensor below 20% is labeled "Activate Sprinkler" and others are "Monitor".

Calculate the average humidity of only the sensors that are not in Danger Zones.

### Հայերեն: Անտառն ունի խոնավության 100 սենսոր: Ցածր խոնավությունը մատնանշում է հրդեհի բարձր վտանգ:

Ստեղծեք խոնավության 100 տոկոսների (0%-ից 100%) զանգված՝ օգտագործելով np.random.randint:

Օգտագործեք համեմատության օպերատորներ՝ «Վտանգավոր գոտիների» համար բուլյան դիմակ (boolean mask) ստեղծելու համար, որտեղ խոնավությունը ցածր է 20%-ից:

Հաշվեք, թե քանի սենսոր է գտնվում վտանգավոր գոտիներում:

Օգտագործեք np.where՝ ստեղծելու նոր զանգված, որը կոչվում է action_plan, որտեղ 20%-ից ցածր ցանկացած սենսոր կնշվի որպես «Ակտիվացնել ցողիչը», իսկ մյուսները՝ «Հսկել»:

Հաշվեք միայն այն սենսորների միջին խոնավությունը, որոնք չեն գտնվում վտանգավոր գոտիներում:

```python
import numpy as np

# 1. 100 Humidity sensors / 100 խոնավության սենսորներ
humidity = np.random.randint(0, 101, 100)

# 2. Boolean mask for Danger Zones / Բուլյան դիմակ վտանգավոր գոտիների համար
danger_mask = humidity < 20

# 3. Count Danger Zones / Հաշվել վտանգավոր գոտիները
# np.sum treats True as 1 and False as 0
danger_count = np.sum(danger_mask)

# 4. Action Plan using np.where / Գործողությունների պլան՝ օգտագործելով np.where
# Syntax: np.where(condition, value_if_true, value_if_false)
action_plan = np.where(humidity < 20, "Activate Sprinkler", "Monitor")

# 5. Average of Safe Zones / Ապահով գոտիների միջին խոնավությունը
# We use the inverse of the mask using the ~ (NOT) operator
safe_avg = humidity[~danger_mask].mean()

print(f"Danger Zones Found: {danger_count}")
print(f"Average Safe Humidity: {safe_avg:.2f}%")
```

## Problem 3: The Broken Fitness Tracker / Կոտրված Ֆիթնես Թրեքերը
English: A user's fitness tracker recorded steps for 50 days, but it malfunctioned and recorded some days as negative numbers or very large outliers (over 50,000 steps).

Create an array of 50 integers between -5000 and 60000.

Clean the data: Any value below 0 or above 30,000 should be treated as invalid and converted to np.nan.

Use np.nanmean to find the user's true daily average.

Calculate the total steps taken over the 50 days, treating NaN values as 0.

Identify the index of the day with the highest valid step count using np.nanargmax.

### Հայերեն: Օգտատիրոջ ֆիթնես թրեքերը 50 օրվա ընթացքում քայլեր է գրանցել, սակայն այն խափանվել է և որոշ օրեր գրանցել է որպես բացասական թվեր կամ շատ մեծ շեղումներ (50,000 քայլից ավելի):

Ստեղծեք 50 ամբողջ թվերի զանգված -5000-ից 60000-ի միջակայքում:

Մաքրեք տվյալները. 0-ից ցածր կամ 30,000-ից բարձր ցանկացած արժեք պետք է դիտարկվի որպես անվավեր և փոխակերպվի np.nan-ի:

Օգտագործեք np.nanmean՝ օգտատիրոջ իրական օրական միջինը գտնելու համար:

Հաշվեք 50 օրվա ընթացքում կատարված ընդհանուր քայլերը՝ NaN արժեքները դիտարկելով որպես 0:

Օգտագործելով np.nanargmax, գտեք այն օրվա ինդեքսը, որն ունի վավեր քայլերի ամենաբարձր քանակը:

```python
import numpy as np

# 1. 50 days of noisy step data / 50 օրվա քայլերի «աղմկոտ» տվյալներ
steps = np.random.randint(-5000, 60000, 50)

# 2. Clean outliers / Մաքրել շեղումները
# Values < 0 or > 30,000 are set to NaN
# We must convert the array to float first because NaN is a float type
steps = steps.astype(float)
steps[(steps < 0) | (steps > 30000)] = np.nan

# 3. True daily average / Իրական օրական միջինը
true_avg = np.nanmean(steps)

# 4. Total steps (treating NaN as 0) / Ընդհանուր քայլերը (NaN = 0)
# np.nan_to_num converts NaN to 0 by default
total_steps = np.sum(np.nan_to_num(steps))
# total_steps = np.nansum(steps)

# 5. Index of highest valid day / Ամենաբարձր վավեր օրվա ինդեքսը
best_day_idx = np.nanargmax(steps)

print(f"Cleaned Average: {true_avg:.0f} steps")
print(f"Total Steps: {total_steps}")
print(f"Best Day Index: {best_day_idx}")
```

## Problem 4: Coffee Shop Peak Hours / Սրճարանի Պիկ Ժամերը
English: You are analyzing sales for 60 minutes across 2 different branches.

Create a 2D array of shape (2, 60) with random sales between $5 and $50.

Calculate the total revenue for each branch using an aggregation.

Use a comparison operator to find during which minutes both branches sold more than $40 simultaneously.

Find the "Peak Minute" (the index where the sum of both branches' sales is the highest).

Use np.diff() to find the minute-to-minute volatility (change in sales) for the first branch.

### Հայերեն: Դուք վերլուծում եք 2 տարբեր մասնաճյուղերի 60 րոպեի վաճառքները:

Ստեղծեք (2, 60) ձևի (shape) 2D զանգված՝ $5-ից $50-ի միջակայքում պատահական վաճառքներով:

Հաշվեք յուրաքանչյուր մասնաճյուղի ընդհանուր հասույթը՝ օգտագործելով ագրեգացիա (aggregation):

Օգտագործեք համեմատության օպերատոր՝ պարզելու համար, թե որ րոպեներին են երկու մասնաճյուղերն էլ միաժամանակ վաճառել $40-ից ավելի:

Գտեք «Պիկ րոպեն» (այն ինդեքսը, որտեղ երկու մասնաճյուղերի վաճառքների գումարը ամենաբարձրն է):

Օգտագործեք np.diff()՝ առաջին մասնաճյուղի րոպե առ րոպե տատանողականությունը (վաճառքի փոփոխությունը) գտնելու համար:

```python
import numpy as np

# 1. 2 Branches, 60 minutes of sales / 2 մասնաճյուղ, 60 րոպեի վաճառք
sales = np.random.randint(5, 51, (2, 60))

# 2. Total revenue per branch / Յուրաքանչյուր մասնաճյուղի ընդհանուր հասույթը
# axis=1 sums horizontally (across the minutes)
branch_totals = sales.sum(axis=1)

# 3. Simultaneous peak minutes / Միաժամանակյա պիկ րոպեներ
# We check where both rows in the same column are > 40
simultaneous_peaks = np.sum((sales[0] > 40) & (sales[1] > 40))

# 4. Peak Minute (Total of both branches) / Պիկ րոպեն (երկու մասնաճյուղի գումարը)
# axis=0 sums vertically (combining both branches for each minute)
minute_totals = sales.sum(axis=0)
peak_minute_idx = np.argmax(minute_totals)

# 5. Volatility (Minute-to-minute change) / Տատանողականություն
# np.diff calculates the difference between adjacent elements
volatility = np.diff(sales[0])

print(f"Branch Totals: Branch A: ${branch_totals[0]}, Branch B: ${branch_totals[1]}")
print(f"Minutes where both sold >$40: {simultaneous_peaks}")
print(f"Highest combined sales minute: {peak_minute_idx}")
```

### Problem 5: Global Shipping Logistics / Համաշխարհային Բեռնափոխադրման Լոգիստիկա
English: You have 80 packages. You need to calculate shipping costs based on weight and distance.Create two 1D arrays of 80 elements: weights (1kg to 50kg) and distances (10km to 1000km).The base shipping formula is: $Cost = (Weight \times 0.5) + (Distance \times 0.1)$. Calculate this using ufuncs.Due to fuel surcharges, any distance over 500km gets an extra $20 flat fee. Use np.where to apply this logic.Calculate the standard deviation of the shipping costs.Identify all packages that cost more than 1.5 standard deviations above the mean cost (High-Cost Deliveries).
## Հայերեն:
 Դուք ունեք 80 ծանրոց: Ձեզ անհրաժեշտ է հաշվարկել առաքման ծախսերը՝ հիմնվելով քաշի և հեռավորության վրա:Ստեղծեք 80 տարրից բաղկացած երկու 1D զանգված՝ weights (1կգ-ից 50կգ) և distances (10կմ-ից 1000կմ):Առաքման հիմնական բանաձևն է՝ $Ծախս = (Քաշ \times 0.5) + (Հեռավորություն \times 0.1): Հաշվեք սա ufunc-ների միջոցով:Վառելիքի հավելավճարների պատճառով 500 կմ-ից ավելի ցանկացած հեռավորության համար գանձվում է լրացուցից $20 հաստատուն վճար: Օգտագործեք np.where այս տրամաբանությունը կիրառելու համար:Հաշվեք առաքման ծախսերի ստանդարտ շեղումը (standard deviation):Հայտնաբերեք բոլոր այն ծանրոցները, որոնց արժեքը 1.5 ստանդարտ շեղումով բարձր է միջին արժեքից (Բարձր արժեքով առաքումներ):

```python
import numpy as np

# 1. 80 packages weights and distances / 80 ծանրոցի քաշ և հեռավորություն
weights = np.random.uniform(1, 50, 80)
distances = np.random.uniform(10, 1000, 80)

# 2. Base Cost formula / Առաքման հիմնական արժեքը
costs = (weights * 0.5) + (distances * 0.1)

# 3. Distance surcharge / Հեռավորության հավելավճար
# Add $20 if distance > 500km
costs = np.where(distances > 500, costs + 20, costs)

# 4. Standard Deviation / Ստանդարտ շեղում
cost_std = np.std(costs)
cost_mean = np.mean(costs)

# 5. Identify High-Cost Deliveries / Բարձր արժեքով առաքումների հայտնաբերում
# Outliers are defined as > (Mean + 1.5 * STD)
high_cost_mask = costs > (cost_mean + 1.5 * cost_std)
high_cost_packages = np.sum(high_cost_mask)

print(f"Standard Deviation of Costs: ${cost_std:.2f}")
print(f"Number of High-Cost Packages: {high_cost_packages}")
```
