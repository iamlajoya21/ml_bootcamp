## Problem 1: The Stock Market Rebound / Ֆոնդային շուկայի վերականգնումը
English: You are managing a portfolio of 80 different stocks. You have their prices at the start of the month and at the end of the month.

Create two arrays of 80 random floats: start_prices ($10 to $500) and end_prices ($10 to $500).

Calculate the percentage change for each stock.

Identify how many stocks saw a gain of more than 15%.

Create a new array that contains the prices of only those "High Gain" stocks at the end of the month.

Հայերեն: Դուք կառավարում եք 80 տարբեր բաժնետոմսերի փաթեթ: Դուք ունեք դրանց գները ամսվա սկզբին և ամսվա վերջին:

Ստեղծեք 80 պատահական թվերից բաղկացած երկու զանգված՝ start_prices ($10-ից $500) և end_prices ($10-ից $500):

Հաշվեք յուրաքանչյուր բաժնետոմսի գնի փոփոխության տոկոսը:

Պարզեք, թե քանի բաժնետոմս է գրանցել ավելի քան 15% աճ:

Ստեղծեք նոր զանգված, որը կպարունակի միայն այդ «Բարձր աճ» ունեցող բաժնետոմսերի գները ամսվա վերջին:

```python
import numpy as np
np.random.seed(42)

# 1. Generate data
start = np.random.uniform(10, 500, 80)
end = np.random.uniform(10, 500, 80)

# 2. Percentage change formula: ((end - start) / start) * 100
# Փոփոխության բանաձևը
pct_change = ((end - start) / start) * 100

# 3. Identify gains > 15% using boolean masking
# Գտնել 15%-ից ավել աճը բուլյան դիմակի միջոցով
high_gain_mask = pct_change > 15
count_high_gain = np.sum(high_gain_mask)

# 4. Extract end prices for high gainers
# Առանձնացնել վերջնական գները բարձր աճողների համար
high_gain_prices = end[high_gain_mask]

print(f"Stocks with >15% gain: {count_high_gain}")
```

## Problem 2: Emergency Room Triage / Շտապօգնության բաժանմունքի տեսակավորում
English: A hospital has 100 patients in the waiting room. Each patient has a Unique ID (1 to 100) and a Wait Time (minutes).

Create an ID array and a wait time array (random integers from 0 to 300 minutes).

You need to treat the patients with the longest wait times first. Rearrange the Patient IDs based on their wait times, from longest to shortest.

Display the IDs of the top 10 most urgent patients.

Հայերեն: Հիվանդանոցի սպասասրահում կա 100 հիվանդ: Յուրաքանչյուր հիվանդ ունի Եզակի ID (1-ից 100) և Սպասման ժամանակ (րոպեներով):

Ստեղծեք ID-ների զանգված և սպասման ժամանակների զանգված (պատահական ամբողջ թվեր 0-ից 300 րոպե):

Դուք պետք է առաջինը սպասարկեք ամենաերկար սպասած հիվանդներին: Վերադասավորեք Հիվանդների ID-ները՝ ըստ իրենց սպասման ժամանակի՝ ամենաերկարից մինչև ամենակարճը:

Ցուցադրեք 10 ամենահրատապ հիվանդների ID-ները:

```python
import numpy as np

# 1. Setup IDs and wait times
patient_ids = np.arange(1, 101)
wait_times = np.random.uno(0, 301, 100)

# 2. Sort indices by wait time (descending)
# argsort վերադարձնում է ինդեքսները, որոնք կտեսակավորեին զանգվածը
# We use [::-1] to flip it from ascending to descending
urgent_indices = np.argsort(wait_times)[::-1]

# 3. Reorder Patient IDs and get top 10
# Վերադասավորել ID-ները և վերցնել առաջին 10-ը
sorted_ids = patient_ids[urgent_indices]
top_10_urgent = sorted_ids[:10]

print(f"Top 10 Urgent Patient IDs: {top_10_urgent}")
```

## Problem 3: Smart Agriculture / Խելացի գյուղատնտեսություն
English: A farm has 120 soil moisture sensors. Some sensors are malfunctioning and returning NaN.

Create an array of 120 moisture readings (0% to 100%).

Randomly replace 20 readings with np.nan.

Calculate the average moisture level of the farm, ignoring the broken sensors.

Replace all NaN values with the average calculated in the previous step to "fix" the data for a report.

Հայերեն: Ֆերման ունի 120 հողի խոնավության սենսոր: Որոշ սենսորներ խափանվել են և վերադարձնում են NaN:

Ստեղծեք 120 խոնավության ցուցմունքների զանգված (0%-ից 100%):

Պատահականորեն փոխարինեք 20 ցուցմունք np.nan-ով:

Հաշվեք ֆերմայի խոնավության միջին մակարդակը՝ անտեսելով խափանված սենսորները:

Բոլոր NaN արժեքները փոխարինեք նախորդ քայլում հաշվարկված միջինով՝ զեկույցի համար տվյալները «շտկելու» նպատակով:

```python
import numpy as np

# 1. 120 Sensors
moisture = np.random.uniform(0, 100, 120)

# 2. Randomly insert 20 NaNs
nan_idx = np.random.choice(120, 20, replace=False)
moisture[nan_idx] = np.nan

# 3. Average ignoring NaNs
# Հաշվել միջինը՝ անտեսելով NaN-երը
avg_moisture = np.nanmean(moisture)

# 4. Fill NaNs with average
# Լրացնել NaN-երը միջինով
moisture[np.isnan(moisture)] = avg_moisture

print(f"Imputed Average Moisture: {avg_moisture:.2f}%")
```

## Problem 5: Global Supply Chain / Համաշխարհային մատակարարման շղթա
English: A logistics company tracks fuel consumption for 5 different delivery routes over 30 days.

Create a 2D array of shape (5, 30) with random fuel usage (50L to 200L per day).

Calculate the total fuel used by each route over the month.

Identify which day (index) was the most expensive (highest fuel usage) for each specific route.

Calculate the average daily fuel usage for the entire company.

Հայերեն: Լոգիստիկ ընկերությունը հետևում է վառելիքի ծախսին 5 տարբեր առաքման երթուղիների համար 30 օրվա ընթացքում:

Ստեղծեք (5, 30) ձևի 2D զանգված՝ վառելիքի պատահական օգտագործմամբ (օրական 50լ-ից 200լ):

Հաշվեք յուրաքանչյուր երթուղու կողմից ամսվա ընթացքում օգտագործված վառելիքի ընդհանուր քանակը:

Յուրաքանչյուր երթուղու համար պարզեք, թե որ օրն (ինդեքսը) է եղել ամենածախսատարը (վառելիքի ամենաբարձր սպառումը):

Հաշվեք ամբողջ ընկերության վառելիքի օրական միջին սպառումը:

```python
import numpy as np

# 1. 5 routes x 30 days
fuel_data = np.random.randint(50, 201, (5, 30))

# 2. Total per route (sum across axis 1)
# Ընդհանուրը ըստ երթուղու
route_totals = fuel_data.sum(axis=1)

# 3. Peak day for each route
# Ամենածախսատար օրը յուրաքանչյուր երթուղու համար
peak_days = fuel_data.argmax(axis=1)

# 4. Company average
# Ընկերության միջինը
overall_avg = fuel_data.mean()

print(f"Total Fuel per Route: {route_totals}")
print(f"Peak Day Indices: {peak_days}")
print(f"Overall Daily Average: {overall_avg:.2f}L")
```
