English: Create a scatter plot of Population vs Area (sq. mi.). Use a log scale for both axes (plt.xscale('log')). Use plt.annotate to point out and label the specific points for "Russia", "Canada", and "China" to show where these giants sit on the map.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

# Load and Clean
url = "https://raw.githubusercontent.com/bnokoro/Data-Science/master/countries%20of%20the%20world.csv"
df = pd.read_csv(url)

# Fix the decimal issue (commas to dots)
cols = ['Pop. Density (per sq. mi.)', 'Birthrate', 'Deathrate', 'Infant mortality (per 1000 births)', 'Literacy (%)', 'Phones (per 1000)', 'Arable (%)', 'Agriculture', 'GDP ($ per capita)']

for col in cols:
    df[col] = df[col].astype(str).str.replace(',', '.').astype(float)
    

df['Region'] = df['Region'].str.strip()
df['Country'] = df['Country'].str.strip()
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
# Using 'viridis' colormap for high contrast
scatter = plt.scatter(df['GDP ($ per capita)'], df['Literacy (%)'], 
                     c=df['Phones (per 1000)'], cmap='viridis', alpha=0.6)

plt.colorbar(label='Phones per 1000 people')
plt.xlabel('GDP ($ per capita)')
plt.ylabel('Literacy Rate (%)')
plt.title('Relationship between Wealth, Literacy, and Technology')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Create a 1x2 grid of plots
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Birthrate
ax[0].hist(df['Birthrate'], bins=20, color='skyblue', edgecolor='black')
ax[0].set_title('Global Birthrate Distribution')
ax[0].set_xlabel('Births per 1000')
ax[0].set_ylabel('Frequency')

# Plot 2: Deathrate
ax[1].hist(df['Deathrate'], 
           bins=20, color='salmon', edgecolor='black')
ax[1].set_title('Global Deathrate Distribution')
ax[1].set_xlabel('Deaths per 1000')

plt.tight_layout() # Prevents label overlapping
plt.show()
```

```python
# Aggregate data
stats = df.groupby('Region')['GDP ($ per capita)'].agg(['mean', 'std']).dropna()

plt.figure(figsize=(12, 6))
plt.errorbar(stats.index, stats['mean'], yerr=stats['std'], 
             fmt='o', color='darkblue', ecolor='red', capsize=5, markersize=8)

plt.xticks(rotation=30, ha = 'right')
plt.ylabel('Average GDP ($)')
plt.title('Regional GDP Averages with Standard Deviation')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Subsets for the Legend
africa = df[df['Region'] == 'SUB-SAHARAN AFRICA']
europe = df[df['Region'] == 'WESTERN EUROPE']

plt.figure(figsize=(10, 6))

# Background: All countries (light gray)
plt.scatter(df['Arable (%)'], df['Agriculture'], color='gray', alpha=0.3, label='Other Regions')

# # Foreground: Africa vs Europe
plt.scatter(africa['Arable (%)'], africa['Agriculture'], color='green', label='Sub-Saharan Africa')
plt.scatter(europe['Arable (%)'], europe['Agriculture'], color='blue', label='Western Europe')

plt.xlabel('Arable Land (%)')
plt.ylabel('Agriculture (% of Economy)')
plt.title('Agriculture Share of Economy vs. Available Arable Land')
plt.legend(loc='upper right', frameon=False)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Area (sq. mi.)'], df['Population'], alpha=0.5, c='darkgreen')

# 1. Vital: Log Scales (Population and Area are too large for linear scales)
plt.xscale('log')
plt.yscale('log')

# 2. More robust matching
targets = ['Russia', 'Canada', 'China']

for country in targets:
    # We use .str.contains and na=False to find the row safely
    mask = df['Country'].str.contains(country, case=False, na=False)
    row = df[mask]
    
    if not row.empty:
        # We take the first match found [0]
        x_val = row['Area (sq. mi.)'].iloc[0]
        y_val = row['Population'].iloc[0]
        
        plt.annotate(country, 
                     xy=(x_val, y_val),           # Point to the data
                     xytext=(20, 20),             # Offset the text
                     textcoords='offset points',  # How to interpret the offset
                     arrowprops=dict(arrowstyle='->', color='black', lw=1),
                     fontsize=10, 
                     fontweight='bold')
    else:
        print(f"Warning: {country} not found in the dataset!")

plt.xlabel('Land Area (sq. mi.) - Log Scale')
plt.ylabel('Population - Log Scale')
# Using LaTeX for the mathematical representation of the axes
# $$ \log_{10}(\text{Area}) \text{ vs } \log_{10}(\text{Population}) $$
plt.title('Global Scale: Population vs. Land Area')
plt.grid(True, which="both", ls="-", alpha=0.2)
plt.show()
```
