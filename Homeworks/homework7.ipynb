```python
import numpy as np

# 1. Create a 5x3 array filled with 7s
# Note: np.full is the most direct way, or np.ones() * 7
matrix = np.full((5, 3), 7)

# 2. Update 3rd row, 2nd column (Index 2, 1)
matrix[2, 1] = 10

# 3. Print attributes
print("Array:\n", matrix)
print(f"Dimensions: {matrix.ndim}")
print(f"Shape: {matrix.shape}")
print(f"Data Type: {matrix.dtype}")
```


<h2>Problem 2: The Linear Navigator</h2>
Task: 1. Create a 1D array called evens that contains all even integers from 10 to 50 (inclusive). 2. Create a second 1D array called linspace_data that contains 15 equally spaced values between 0 and 100. 3. Calculate and print the total memory size (in bytes) of the evens array.

Hint: Total bytes = .size multiplied by .itemsize.


```python
evens_float = np.arange(10.0, 52.0, 2)
evens_int = np.arange(10, 52, 2)
```

```python


evens = np.array([1,2], dtype='int8')

print(evens_float.itemsize)
print(evens_int.itemsize)
print(evens.itemsize)

```

```python
import numpy as np

# 1. Evens from 10 to 50
evens = np.arange(10, 52, 2)

# 2. 15 equally spaced values from 0 to 100
linspace_data = np.linspace(0, 100, 15)

# 3. Calculate total bytes
total_bytes = evens.size * evens.itemsize

print(f"Evens Array: {evens}")
print(f"Linspace Array: {linspace_data}")
print(f"Total bytes used by 'evens': {total_bytes} bytes")
```

<h2>Problem 3: The 3D Tensor Check</h2>
Task: Imagine you are storing data for a 3-day experiment. Each day, you take 4 readings from 2 different sensors.

Create a 3D array of zeros with the shape (3, 4, 2) to represent this data.

Print the size of the array.

Verify the "Rank" of the array by printing its ndim.

Explain in a comment what each number in the shape (3, 4, 2) represents based on the experiment description.

```python
import numpy as np

# 1. Create 3D zeros (3 days, 4 readings, 2 sensors)
experiment_data = np.empty((3, 4, 2))
print(experiment_data)
# 2. Print size
print(f"Total number of readings (size): {experiment_data.size}")

# 3. Print rank
print(f"Array Rank (ndim): {experiment_data.ndim}")

# 4. Explanation:
# The shape (3, 4, 2) represents:
# 3: The 'depth' or number of blocks (The 3 Days)
# 4: The number of rows in each block (The 4 Readings per day)
# 2: The number of columns (The 2 Sensors per reading)
```

```python
import numpy as np

# 1. Create the array
temps_f = np.array([72, 75, 68, 80, 82, 77, 74, 70, 65, 85])

# 2. Convert to Celsius (Vectorized operation)
temps_c = (temps_f - 32) * (5/9)

# 3. Round to 2 decimal places
temps_c_rounded = np.round(temps_c, 2)

print(f"Fahrenheit: {temps_f}")
print(f"Celsius:    {temps_c_rounded}")
```

```python
import numpy as np

prices = np.array([120, 85, 200, 45, 150])
costs = np.array([80, 60, 130, 30, 100])

# 1. Element-wise subtraction
profits = np.subtract(prices, costs) # or simply: prices - costs

# 2. Square root ufunc
roots = np.sqrt(profits)

# 3. Max ufunc
max_profit = np.max(profits)

print(f"Profits: {profits}")
print(f"Square Roots: {roots}")
print(f"Highest Profit: {max_profit}")
```

```python
import numpy as np

# 1. Generate 10 values between 0 and 2*pi
x = np.linspace(0, 2 * np.pi, 10)

# 2. Calculate Sin and Cos
sin_x = np.sin(x)
cos_x = np.cos(x)

# 3. Sum the signals
result = sin_x + cos_x

print(f"X values: {x}")
print(f"Resulting Signal: {result}")
```
