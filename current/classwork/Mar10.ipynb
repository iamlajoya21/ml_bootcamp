```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 1. Load Data
data = load_wine()
X, y = data.data, data.target

# 2. Create Polynomial Features (Degree 2)
# This adds interaction terms and squared features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# 3. Split Data (80% Train, 20% Test)
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.20, random_state=42)

# 4. Define Hyperparameter Grid
# English: These are the settings GridSearchCV will combine and test.
# Հայերեն: Սրանք այն կարգավորումներն են, որոնք GridSearchCV-ն կհամակցի և կփորձարկի:
param_grid = {
    'n_estimators': [50, 100, 200],          # Number of trees / Ծառերի քանակը
    'max_depth': [None, 5, 10],              # Limit depth to control variance / Սահմանափակել խորությունը
    'min_samples_split': [2, 5, 10],         # Min samples to split a node / Նվազագույն նմուշները ճյուղավորման համար
    'criterion': ['gini', 'entropy']         # Quality of split / Ճյուղավորման որակը
}

# 5. Initialize and Run GridSearchCV
# Using 5-fold cross-validation
rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)

# 6. Evaluation and Results
best_model = grid_search.best_estimator_
test_acc = accuracy_score(y_test, best_model.predict(X_test))
cv_score = grid_search.best_score_

print("-" * 30)
print(f"Best Hyperparameters: {grid_search.best_params_}")
print(f"Mean CV Accuracy (Validation): {cv_score:.4f}")
print(f"Test Set Accuracy: {test_acc:.4f}")
print("-" * 30)

# # 7. Bias-Variance Analysis
# # If CV score is much higher than Test score, we have High Variance.
# variance_gap = abs(cv_score - test_acc)
# if variance_gap > 0.05:
#     print(f"Warning: High Variance detected. Gap: {variance_gap:.4f}")
# else:
#     print("Model is well-generalized (Low Variance gap).")
```
