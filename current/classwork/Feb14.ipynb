```python
import pandas as pd
import ssl

# --- SECTION 0: ENVIRONMENT SETUP ---
# Fix for the SSL certificate verification error
ssl._create_default_https_context = ssl._create_unverified_context

# Dataset URLs
AIRPORT_URL = "https://raw.githubusercontent.com/datasets/airport-codes/refs/heads/main/data/airport-codes.csv"
COUNTRY_URL = "https://raw.githubusercontent.com/datasets/country-codes/master/data/country-codes.csv"

# Load data
print("📡 Loading global datasets...")
df_air = pd.read_csv(AIRPORT_URL)
df_geo = pd.read_csv(COUNTRY_URL)

# --- SECTION 1: DATA LINKAGE (MERGE) ---
# We use a LEFT merge to ensure we don't lose airports that might 
# not have matching country metadata in our reference file.
df = pd.merge(
    df_air, 
    df_geo, 
    left_on='iso_country', 
    right_on='ISO3166-1-Alpha-2', 
    how='left'
)

# --- SECTION 2: CONTINENT AUDIT ---
# Identifying discrepancies between the two datasets. 
# We use .fillna() to allow comparisons even where data is missing.
mismatches = df[df['continent'].fillna('NA') != df['Continent'].fillna('NA')]

# --- SECTION 3: GEOGRAPHICAL FEATURE ENGINEERING ---
# The 'coordinates' column is a string. We need to split it and convert to numeric floats.

coords_split = df['coordinates'].str.split(',', expand=True)
df['Longitude'] = coords_split[0].astype(float)
df['Latitude'] = coords_split[1].astype(float)

# --- SECTION 4: THE AVIATION CEILING REPORT ---
# 1. Filter for major infrastructure (Large and Medium airports)
filtered_infra = df[df['type'].isin(['large_airport', 'medium_airport'])]

# 2. Find the top 2 highest airports per continent.
# By using group_keys=False, we keep the structure cleaner for the result.
aviation_ceiling = filtered_infra.groupby('continent', group_keys=False).apply(
    lambda x: x.nlargest(2, 'elevation_ft')
)

# Reset index to ensure 'continent' is a column and not a hidden index label
# This prevents the KeyError: "['continent'] not in index"
aviation_ceiling = aviation_ceiling.reset_index()

ceiling_report = aviation_ceiling[['name', 'official_name_en', 'elevation_ft', 'continent']]

# --- SECTION 5: INFRASTRUCTURE DENSITY ANALYSIS ---
# Calculate density using the formula:
# $$Density = \frac{Total Airports}{Population} \times 1,000,000$$

# Count airports per country
airport_counts = df.groupby('iso_country').size().reset_index(name='total_airports')

# Link counts back to population data
density_df = pd.merge(
    airport_counts, 
    df_geo[['ISO3166-1-Alpha-2', 'official_name_en', 'Population']], 
    left_on='iso_country', 
    right_on='ISO3166-1-Alpha-2'
)

# Calculate the density metric
density_df['Density_per_Million'] = (density_df['total_airports'] / density_df['Population']) * 1_000_000

# Get the top 5 ranking countries
top_5_density = density_df.sort_values(by='Density_per_Million', ascending=False).head(5)

# --- FINAL OUTPUTS ---
print("\n" + "="*50)
print("📊 FINAL REPORT: AVIATION EXTREMES")
print("="*50)

print(f"\n1. AUDIT: Found {len(mismatches)} continent discrepancies.")

print("\n2. TOP 2 HIGHEST AIRPORTS PER CONTINENT:")
print(ceiling_report.to_string(index=False))

print("\n3. TOP 5 COUNTRIES BY AIRPORT DENSITY (Per 1M People):")
print(top_5_density[['official_name_en', 'total_airports', 'Density_per_Million']].to_string(index=False))
```
