```python

# --- Solution 8: Friday Future Return ---
fridays = goog[goog.index.dayofweek == 4].copy()
# We look for the price 3 days after Friday (Monday)
# method='bfill' handles weekends/holidays automatically
monday_prices = goog['Close'].reindex(fridays.index + pd.Timedelta(days=3), method='bfill')

# Calculate return between Friday Close and Monday (bfilled) Close
fridays['monday_return'] = (monday_prices.values - fridays['Close'].values) / fridays['Close'].values
```
