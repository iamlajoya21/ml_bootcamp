```python
import pandas as pd
import numpy as np

# Assuming df is already loaded with your lowercase columns
# df = sns.load_dataset('titanic')

# --- Solution 1: Survival Matrix ---
# Pivot table defaults to 'mean' for numeric values
ans_1 = df.pivot_table(index='sex', columns='pclass', values='survived')

# --- Solution 2: Fare & Port Logistics ---
# We use embark_town instead of embarked for better readability
ans_2 = df.pivot_table(
    index='embark_town', 
    columns='pclass', 
    values='fare', 
    aggfunc=['sum', 'mean']
)

# --- Solution 3: Demographic Overview ---
# We use 'pclass' to count the total passengers (it's non-null)
ans_3 = df.pivot_table(
    index=['sex', 'survived'], 
    values=['age', 'pclass'], 
    aggfunc={'age': 'mean', 'pclass': 'count'}
)
# Rename 'pclass' in the result to 'passenger_count' for clarity
ans_3 = ans_3.rename(columns={'pclass': 'passenger_count'})
```
