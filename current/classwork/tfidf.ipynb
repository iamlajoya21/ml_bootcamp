```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download NLTK data (run once)
nltk.download('stopwords')
nltk.download('wordnet')

df_movies = pd.read_csv(movie_file_path)

# Let's filter to just a few top genres to make classification easier for your first run
top_genres = ['comedy', 'drama', 'action', 'thriller', 'romance']
df_movies = df_movies[df_movies['Genre'].isin(top_genres)]

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Lowercase
    text = str(text).lower()
    # Remove punctuation and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenize and remove stop words + lemmatize
    tokens = text.split()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)

# Apply the preprocessing
df_movies['Clean_Plot'] = df_movies['Plot'].apply(preprocess_text)
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Initialize and fit the TF-IDF Vectorizer
# We limit max_features to keep it computationally light and use english stop words
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df_movies['Clean_Plot'])

# 2. Extract the vocabulary (the actual words)
feature_names = np.array(vectorizer.get_feature_names_out())

# 3. Create a helper function to grab the top words row by row
def get_top_keywords_for_all(tfidf_matrix, features, top_n=5):
    all_keywords = []
    
    # Iterate through each document (row) in the sparse matrix
    for row in tfidf_matrix:
        # Convert the single row to a dense array so we can sort it
        row_data = row.toarray().flatten()
        
        # Get the indices of the highest scores (argsort sorts ascending, so we take the end and reverse)
        top_indices = row_data.argsort()[-top_n:][::-1]
        
        # Map indices back to the actual words, ignoring words with a score of 0
        top_words = [features[i] for i in top_indices if row_data[i] > 0]
        
        # Join them into a single string (e.g., "alien, space, ship") and add to our list
        all_keywords.append(", ".join(top_words))
        
    return all_keywords

# 4. Apply the function and create a new column in your DataFrame
df_movies['Top_Keywords'] = get_top_keywords_for_all(tfidf_matrix, feature_names, top_n=5)

# 5. Check out the results!
df_movies[['Clean_Plot', 'Top_Keywords']].head()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Assuming you have your train/test split ready from the articles dataset
# X_train_art, X_test_art, y_train_art, y_test_art 

# 1. Set up the Pipeline
# We can use TfidfVectorizer directly instead of combining CountVectorizer + TfidfTransformer
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english')),
    ('clf', LogisticRegression(max_iter=1000, random_state=42))
])

# 2. Define the Parameter Grid
# Note the syntax: stepname__parametername (with a double underscore in between)
param_grid = {
    # Test Unigrams alone, Unigrams + Bigrams, and Unigrams + Bigrams + Trigrams
    'tfidf__ngram_range': [(1, 1), (1, 2), (1, 3)], 
    
    # Test different vocabulary limits (None means it will use all available features)
    'tfidf__max_features': [1000, 3000, 5000, None],
    
    # Optional: You can even test Logistic Regression hyperparameters at the same time!
    # 'clf__C': [0.1, 1.0, 10.0] 
}

# 3. Initialize the Grid Search
# cv=3 means 3-fold cross-validation. 
# n_jobs=-1 tells your machine to use all available CPU cores to speed up the process.
grid_search = GridSearchCV(pipeline, param_grid, cv=3, n_jobs=-1, verbose=1)

# 4. Run the Grid Search
print("Starting Grid Search... (This will fit multiple models, so it may take a moment!)")
grid_search.fit(X_train, y_train)

# 5. Evaluate the Results
print("\n--- Grid Search Complete ---")
print("Best Parameters found:")
for param, value in grid_search.best_params_.items():
    print(f"  - {param}: {value}")
    
print(f"Best Cross-Validation Accuracy: {grid_search.best_score_:.4f}\n")

# 6. Test the best model on your unseen test data
best_model = grid_search.best_estimator_
final_predictions = best_model.predict(X_test)

print("Classification Report using the Best Model:")
print(classification_report(y_test, final_predictions))
```

```python
from tqdm import tqdm
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Assumptions: X_train_art, X_test_art, y_train_art, y_test_art are already defined

# The specific feature sizes you want to test
feature_sizes = [200, 500, 1000, 2500, 5000, 10000, 20000, 50000, 100000]

# List to store our results for easy viewing later
results = []

# Wrap the list in tqdm() to generate the progress bar
for size in tqdm(feature_sizes, desc="Evaluating max_features"):
    
    # 1. Initialize the Vectorizer and Classifier for the current iteration
    # (Keeping ngram_range=(1, 2) based on our previous setup for better context)
    vectorizer = TfidfVectorizer(max_features=size, stop_words='english', ngram_range=(1, 2))
    clf = LogisticRegression(max_iter=1000, random_state=42)
    
    # 2. Fit and transform the training data, but ONLY transform the test data
    X_train_vec = vectorizer.fit_transform(X_train)
    X_test_vec = vectorizer.transform(X_test)
    
    # 3. Train the Logistic Regression model
    clf.fit(X_train_vec, y_train)
    
    # 4. Make predictions on the test set
    predictions = clf.predict(X_test_vec)
    
    # 5. Calculate accuracy and store it
    acc = accuracy_score(y_test, predictions)
    results.append((size, acc))

# Print the final summary once the loop is completely finished
print("\n--- Results Summary ---")
for size, acc in results:
    print(f"Max Features: {size:<5} | Accuracy: {acc:.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, adjusted_rand_score

# =====================================================================
# 0. PREPARATION & ASSUMPTIONS
# =====================================================================
# Assuming df_movies and df_articles are already loaded and preprocessed 
# from our previous steps.
# Movies text: df_movies['Clean_Plot'] | Target: df_movies['Genre']
# Articles text: df_articles['content'] (Let's assume it's preprocessed into 'Clean_Content') | Target: df_articles['category']

print("Initializing Advanced Text Analytics Pipeline...")

# Let's vectorize both datasets using a researcher-grade TF-IDF setup
# (Removing corpus-specific stop words and limiting features for computational efficiency)
movie_vectorizer = TfidfVectorizer(max_features=3000, min_df=2, max_df=0.95)
X_movies_tfidf = movie_vectorizer.fit_transform(df_movies['Clean_Plot'])

article_vectorizer = TfidfVectorizer(max_features=3000, min_df=2, max_df=0.95)
X_articles_tfidf = article_vectorizer.fit_transform(df_articles['content'])


# =====================================================================
# TASK 1A: DOCUMENT SIMILARITY (Information Retrieval)
# =====================================================================
def get_top_similar_documents(tfidf_matrix, dataframe, text_col, doc_index, top_n=3):
    """
    Computes pairwise cosine similarity between a target document and the rest of the corpus.
    """
    # Compute similarity of the selected document against ALL documents in the matrix
    cosine_sim = cosine_similarity(tfidf_matrix[doc_index], tfidf_matrix).flatten()
    
    # Sort indices by highest similarity score (excluding the document itself, which is index -1 after sorting)
    similar_indices = cosine_sim.argsort()[-(top_n + 1):-1][::-1]
    
    print(f"\n--- Target Document (Index {doc_index}) ---")
    print(f"Original Text snippet: {dataframe[text_col].iloc[doc_index][:150]}...\n")
    print(f"Top {top_n} Most Similar Documents:")
    
    for idx in similar_indices:
        score = cosine_sim[idx]
        print(f"- [Doc Index {idx}] (Similarity Score: {score:.4f}): {dataframe[text_col].iloc[idx][:120]}...")

# Execute Similarity Diagnostics
print("\n=== RUNNING DOCUMENT SIMILARITY DISCOVERY ===")
get_top_similar_documents(X_movies_tfidf, df_movies, 'Plot', doc_index=42, top_n=2)
get_top_similar_documents(X_articles_tfidf, df_articles, 'content', doc_index=10, top_n=2)


# =====================================================================
# TASK 1B: DOCUMENT CLUSTERING & UNSUPERVISED EVALUATION
# =====================================================================
def perform_clustering_analysis(tfidf_matrix, dataframe, target_col, num_clusters, vectorizer):
    """
    Clusters documents using K-Means, maps clusters back to text features, 
    and validates cluster purity using mathematical metrics.
    """
    # Fit K-Means
    kmeans = KMeans(n_clusters=num_clusters, init='k-means++', max_iter=300, random_state=42)
    cluster_labels = kmeans.fit_predict(tfidf_matrix)
    
    # Mathematical Evaluation
    sil_score = silhouette_score(tfidf_matrix, cluster_labels, sample_size=2000, random_state=42)
    # Adjusted Rand Index measures agreement between cluster assignments and true labels (-1 to 1)
    ari_score = adjusted_rand_score(dataframe[target_col], cluster_labels)
    
    print(f"\n--- Clustering Metrics (K={num_clusters}) ---")
    print(f"Mean Silhouette Coefficient: {sil_score:.4f} (Closer to 1 = distinct, well-separated clusters)")
    print(f"Adjusted Rand Index (ARI): {ari_score:.4f} (Closer to 1 = perfectly matches ground-truth labels)")
    
    # Top Terms Per Cluster (Centroid Analysis)
    print("\nTop Terms per Cluster Centroid:")
    order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
    terms = vectorizer.get_feature_names_out()
    
    for i in range(num_clusters):
        top_terms = [terms[ind] for ind in order_centroids[i, :7]]
        print(f"  Cluster {i}: {', '.join(top_terms)}")
        
    return cluster_labels

# Execute Clustering (Using 5 clusters for both since they roughly map to our 5 target classes)
print("\n=== RUNNING UNSUPERVISED MOVIE CLUSTERING ===")
df_movies['Cluster'] = perform_clustering_analysis(X_movies_tfidf, df_movies, 'Genre', num_clusters=5, vectorizer=movie_vectorizer)

print("\n=== RUNNING UNSUPERVISED ARTICLE CLUSTERING ===")
df_articles['Cluster'] = perform_clustering_analysis(X_articles_tfidf, df_articles, 'category', num_clusters=5, vectorizer=article_vectorizer)


# =====================================================================
# TASK 2: KEYWORD EXTRACTION via TF-IDF Weight Matrix
# =====================================================================
def extract_tfidf_keywords(tfidf_matrix, vectorizer, doc_index, top_n=5):
    """
    Extracts the highest-weighted terms unique to a specific document inside the corpus.
    """
    feature_names = vectorizer.get_feature_names_out()
    
    # Get the row vector for the specific document
    doc_vector = tfidf_matrix[doc_index].toarray().flatten()
    
    # Sort indices by highest TF-IDF score
    top_keyword_indices = doc_vector.argsort()[-top_n:][::-1]
    
    # Return terms paired with their mathematical weights
    keywords = [(feature_names[i], doc_vector[i]) for i in top_keyword_indices if doc_vector[i] > 0]
    return keywords

# Execute Keyword Extraction Diagnostics
print("\n=== RUNNING DETERMINISTIC KEYWORD EXTRACTION ===")
movie_idx = 105
movie_kw = extract_tfidf_keywords(X_movies_tfidf, movie_vectorizer, doc_index=movie_idx)
print(f"\nTop Keywords for Movie Index {movie_idx} (Plot: {df_movies['Plot'].iloc[movie_idx][:100]}...):")
for word, score in movie_kw:
    print(f"  - {word}: {score:.4f}")

article_idx = 250
article_kw = extract_tfidf_keywords(X_articles_tfidf, article_vectorizer, doc_index=article_idx)
print(f"\nTop Keywords for Article Index {article_idx} (Content: {df_articles['content'].iloc[article_idx][:100]}...):")
for word, score in article_kw:
    print(f"  - {word}: {score:.4f}")
```
