## Problem 1  The Leaderboard / Առաջատարների աղյուսակը
English: You have the results of a 50-meter dash for 60 athletes.
Create an array of 60 random float values between 9.5 and 15.0 representing race times in seconds.
Use np.sort to display the times from fastest to slowest.Use a slicing technique on the sorted array to extract the 3 fastest "Medalist" times.Հայերեն: Դուք ունեք 60 մարզիկների 50 մետր վազքի արդյունքները:
Ստեղծեք 60 պատահական լողացող ստորակետով թվերի զանգված 9.5-ից 15.0-ի միջակայքում, որոնք ներկայացնում են վազքի ժամանակները վայրկյաններով:
Օգտագործեք np.sort՝ ժամանակները ամենաարագից մինչև ամենադանդաղը ցուցադրելու համար:Օգտագործեք սլայսինգի (slicing) տեխնիկան տեսակավորված զանգվածի վրա՝ լավագույն 3 «Մեդալակիրների» ժամանակները առանձնացնելու համար:[8:52 PM]

```python
import numpy as np

# 1. Generate 60 race times / Ստեղծել 60 վազքի ժամանակներ
np.random.seed(1)
times = np.random.uniform(9.5, 15.0, 60)

# 2. Full sort (fastest to slowest) / Ամբողջական տեսակավորում (արագից դանդաղ)
sorted_times = np.sort(times)

# 3. Slice the top 3 / Առանձնացնել լավագույն 3-ը
medalists = sorted_times[:3]

print(f"Fastest 3 Times: {medalists}")
```

## Problem 2 The Inventory Linker / Պահեստի կապակցված գույքագրումը
English: You are managing a store. You have two separate arrays: one for Product IDs (1001 to 1050) and one for their Remaining Stock (50 random integers).
Generate the ID array and the stock array.Find the sequence of positions that would put the stock levels in order from least to most.Use that sequence to display the IDs and their stock levels together, ensuring the product with the lowest stock appears first.Հայերեն: Դուք կառավարում եք խանութ: Դուք ունեք երկու առանձին զանգված՝ մեկը Ապրանքի ID-ների համար (1001-ից 1050), մյուսը՝ դրանց Մնացորդային քանակի (50 պատահական ամբողջ թվեր):
Ստեղծեք ID-ների զանգվածը և քանակների զանգվածը:Գտեք դիրքերի այն հաջորդականությունը, որը քանակները կդասավորի նվազագույնից առավելագույնը:Օգտագործեք այդ հաջորդականությունը՝ ID-ները և դրանց քանակները միասին ցուցադրելու համար՝ ապահովելով, որ ամենաքիչ քանակով ապրանքը հայտնվի առաջինը:

```python
import numpy as np

# 1. Setup IDs and Stock / Ստեղծել ID-ներ և քանակներ
ids = np.arange(1001, 1051)
stock = np.random.randint(0, 500, 50)

# 2. Get the indices that WOULD sort the stock / Ստանալ այն ինդեքսները, որոնք կտեսակավորեին քանակը
# This doesn't move the stock, it just gives us a 'map'
sort_map = np.argsort(stock)

# 3. Use the map to reorder both arrays / Օգտագործել քարտեզը երկու զանգվածները վերադասավորելու համար
sorted_ids = ids[sort_map]
sorted_stock = stock[sort_map]

# Display first 5 for verification / Ցուցադրել առաջին 5-ը ստուգման համար
print("Sorted Inventory (Lowest Stock First):")
for i in range(5):
    print(f"ID: {sorted_ids[i]} | Stock: {sorted_stock[i]}")
```

English: You are a climate scientist monitoring 10 different cities over a period of 50 days (a 500-point data grid).
Create a 2D array of shape (10, 50) containing random temperatures between -30 and 45.
For each city individually, isolate the 3 coldest days in its history. You must do this efficiently—do not perform a full sort of the 50-day history for any city.
Identify the absolute hottest day recorded in the entire 500-point dataset. Instead of just the temperature, return the exact City Index and Day Index where this heatwave occurred.
Create a "Masked Anomaly Map": Generate a new 2D array where only the 3 coldest days for each city retain their original temperature, and all other 47 days per city are replaced with np.nan.Հայերեն: Դուք կլիմայագետ եք, ով վերահսկում է 10 տարբեր քաղաքներ 50 օրվա ընթացքում (500 կետանոց տվյալների ցանց):

Հայտնաբերեք ամբողջ 500 կետանոց տվյալների բազայում գրանցված բացարձակ ամենաշոգ օրը: Միայն ջերմաստիճանի փոխարեն վերադարձրեք այն Քաղաքի ինդեքսը և Օրվա ինդեքսը, որտեղ տեղի է ունեցել այդ շոգի ալիքը:
Ստեղծեք «Անոմալիաների դիմակավորված քարտեզ». ստեղծեք նոր 2D զանգված, որտեղ միայն յուրաքանչյուր քաղաքի 3 ամենացածր ջերմաստիճանով օրերը կպահպանեն իրենց սկզբնական ջերմաստիճանը, իսկ յուրաքանչյուր քաղաքի մյուս 47 օրերը կփոխարինվեն np.nan-ով:

```python
import numpy as np

# 1. 10 cities x 50 days / 10 քաղաք x 50 օր
# np.random.seed(42)
data = np.random.randint(-30, 46, (10, 50))

# 2. Isolate 3 coldest days per city / Առանձնացնել յուրաքանչյուր քաղաքի 3 ամենացածր ջերմաստիճանները
# Use argpartition across axis 1 to get indices of 3 smallest values per row
coldest_indices = np.argpartition(data, kth=3, axis=1)[:, :3]

# 3. Absolute hottest day coordinate / Բացարձակ ամենաշոգ օրվա կոորդինատը
# argmax on a flattened array, then unravel to 2D
flat_max_idx = np.argmax(data)
city_idx, day_idx = np.unravel_index(flat_max_idx, data.shape)

# 4. Masked Anomaly Map / Անոմալիաների դիմակավորված քարտեզ
# Create an empty NaN array
anomaly_map = np.full(data.shape, np.nan)

# Use advanced indexing to fill only the coldest spots
# We need row indices matching the shape of coldest_indices
row_indices = np.arange(10).reshape(-1, 1)
anomaly_map[row_indices, coldest_indices] = data[row_indices, coldest_indices]

print(f"Absolute Heatwave: {data[city_idx, day_idx]}°C at City {city_idx}, Day {day_idx}")
print("\nAnomaly Map (First City - first 10 days):")
print(anomaly_map)
```
