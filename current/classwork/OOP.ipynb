````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Defines behavior for the '+' operator
    def __add__(self, other):
        if isinstance(other, Point):
            return Point(self.x + other.x, self.y + other.y)
        return NotImplemented

    # Defines behavior for the '-' operator
    def __sub__(self, other):
        if isinstance(other, Point):
            return Point(self.x - other.x, self.y - other.y)
        return NotImplemented

    # Defines how the object prints as a string
    def __str__(self):
        return f"Point({self.x}, {self.y})"

# Usage
p1 = Point(1, 4)
p2 = Point(2, 3)

# Uses __add__ internally
p3 = p1 + p2
print(p3)  # Output: Point(3, 7)

# Uses __sub__ internally
p4 = p1 - p2
print(p4)  # Output: Point(-1, 1)
````

````python
import torch 
import torch.nn as nn
# 8. A Custom PyTorch Layer (State + Behavior)
class CustomLinear(nn.Module):
    def __init__(self):
        super().__init__()
        # nn.Parameter tells PyTorch to track these state variables for training
        self.weight = nn.Parameter(torch.tensor([2.0]))
        self.bias = nn.Parameter(torch.tensor([1.0]))
        
    def forward(self, x):
        return (self.weight * x) + self.bias

my_layer = CustomLinear()
# In PyTorch, calling the object directly runs .forward()
print(f"Example 8: CustomLinear output: {my_layer(torch.tensor([3.0]))}")
````

````python

# 9. Composition (A Real PyTorch Neural Network)

import torch
import torch.nn as nn
import torch.optim as optim

# ==========================================
# PHASE 1: PURE PYTHON OOP BASICS
# ==========================================

# --- Example 1: Classes and Objects ---
print("--- Example 1: Classes and Objects ---")
class SimpleNeuron:
    pass

my_neuron = SimpleNeuron()
print("Type of my_neuron:", type(my_neuron))


# --- Example 2: State (The __init__ Method) ---
print("\n--- Example 2: State ---")
class NeuronWithState:
    def __init__(self, weight, bias):
        self.weight = weight
        self.bias = bias

neuron_A = NeuronWithState(weight=0.5, bias=0.1)
print("Neuron A weight:", neuron_A.weight)


# --- Example 3: Behavior (Methods) ---
print("\n--- Example 3: Behavior ---")
class ActiveNeuron:
    def __init__(self, weight, bias):
        self.weight = weight
        self.bias = bias
        
    def forward(self, x):
        return (self.weight * x) + self.bias

n = ActiveNeuron(weight=2.0, bias=1.0)
print("ActiveNeuron forward(3.0):", n.forward(x=3.0))


# --- Example 4: Inheritance ---
print("\n--- Example 4: Inheritance ---")
class ReluNeuron(ActiveNeuron):
    def forward(self, x):
        raw_output = super().forward(x)
        return max(0.0, raw_output)

relu_n = ReluNeuron(weight=-2.0, bias=1.0)
print("ReluNeuron forward(3.0):", relu_n.forward(3.0))


#``` --- Example 5: Composition ---
print("\n--- Example 5: Composition ---")
class SimpleNetwork:
    def __init__(self):
        self.neuron1 = ReluNeuron(weight=1.5, bias=0.5)
        self.neuron2 = ReluNeuron(weight=-0.5, bias=2.0)
        
    def forward(self, x):
        out1 = self.neuron1.forward(x)
        out2 = self.neuron2.forward(out1)
        return out2

my_net = SimpleNetwork()
print("SimpleNetwork forward(2.0):", my_net.forward(2.0))


# ==========================================
# PHASE 2: OOP APPLIED IN PYTORCH
# ==========================================

# --- Example 6: Tensors are Objects Too ---
print("\n--- Example 6: Tensors as Objects ---")
x = torch.tensor([1.0, -2.0, 3.0])
print("Tensor shape:", x.shape)
print("Tensor mean:", x.mean().item())
print("Tensor relu:", x.relu())


# --- Example 7: The Foundation (nn.Module) ---
print("\n--- Example 7: The Foundation (nn.Module) ---")
class BlankModel(nn.Module):
    def __init__(self):
        super().__init__() 

blank_model = BlankModel()
print("BlankModel instantiated successfully.")


# --- Example 8: A Custom PyTorch Layer ---
print("\n--- Example 8: A Custom PyTorch Layer ---")
class CustomLinear(nn.Module):
    def __init__(self):
        super().__init__()
        self.weight = nn.Parameter(torch.tensor([2.0]))
        self.bias = nn.Parameter(torch.tensor([1.0]))
        
    def forward(self, x):
        return (self.weight * x) + self.bias

my_layer = CustomLinear()
# Calling the object directly runs the forward() method
print("CustomLinear output:", my_layer(torch.tensor([3.0])))


# --- Example 9: Composition (A Real Neural Network) ---
print("\n--- Example 9: Composition (Real Network) ---")
class MultiLayerPerceptron(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden_layer = nn.Linear(in_features=10, out_features=5)
        self.output_layer = nn.Linear(in_features=5, out_features=1)
        self.activation = nn.ReLU()
        
    def forward(self, x):
        x = self.hidden_layer(x)
        x = self.activation(x)
        x = self.output_layer(x)
        return x

model = MultiLayerPerceptron()
print("Model architecture:\n", model)


# --- Example 10: Objects Interacting (The Training Loop) ---
print("\n--- Example 10: The Training Loop ---")
loss_function = nn.MSELoss() 
optimizer = optim.SGD(model.parameters(), lr=0.01) 

# Dummy data
inputs = torch.randn(1, 10)
targets = torch.tensor([[1.0]])

print("Starting training loop...")
for epoch in range(3):
    optimizer.zero_grad()            
    
    predictions = model(inputs)      
    loss = loss_function(predictions, targets) 
    loss.backward()                  
    
    optimizer.step()                 
    
    print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
````
