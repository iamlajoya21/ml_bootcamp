```python
# Solution 1: The Department Grouper

# Problem 1: The Coordinate Generator (Nested For)


# Task: You are building a game board. 
# Create a generic list called grid and fill it with tuples representing coordinates (x, y).
# Iterate x from 0 to 2.
# Inside that, iterate y from 0 to 2.
# Result: [(0, 0), (0, 1), (0, 2), (1, 0), ...]


employees = ["Alice", "Bob", "Charlie", "David", "Eve"]
depts = ["IT", "HR", "IT", "Sales", "HR"]

org_chart = {}

for emp, d in zip(employees, depts):
    # Check if this department already exists in our dictionary
    if d in org_chart:
        org_chart[d].append(emp)
    else:
        # If not, create a NEW list with this employee in it
        org_chart[d] = [emp]

import pprint
pprint.pprint(org_chart)
```

```python
names = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace"]
depts = ["IT", "HR", "Sales", "IT", "HR", "Intern", "IT"]
locs  = ["NY", "NY", "NY", "LA", "LA", "NY", "NY"]

directory = {}

# Zip all three lists together
for name, dept, loc in zip(names, depts, locs):
    
    # Logic Step 1: Filter out Interns
    if dept == "Intern":
        continue
    
    # Logic Step 2: Ensure the Location (Level 1) exists
    if loc not in directory:
        directory[loc] = {} # Initialize an empty dictionary for this city
        
    # Logic Step 3: Ensure the Department (Level 2) exists INSIDE that Location
    # Note: We access directory[loc] first!
    if dept not in directory[loc]:
        directory[loc][dept] = [] # Initialize an empty list for this dept
        
    # Logic Step 4: Add the name
    directory[loc][dept].append(name)

# Pretty print to see the structure
import pprint
pprint.pprint(directory)
```

```python
# Solution 2: The "Curve" Calculator





students = ["Sam", "Jess", "Tom"]
raw_scores = [80, 40, 75]

# Pass 1: Find the highest score to calculate the curve base
highest = 0
for s in raw_scores:
    if s > highest:
        highest = s

final_grades = {}

# Pass 2: Calculate grades
for name, score in zip(students, raw_scores):
    # Avoid division by zero if highest is 0
    if highest > 0:
        curved_score = (score / highest) * 100
        final_grades[name] = curved_score
    else:
        final_grades[name] = 0

print(final_grades)
```

```python
# Solution 3: The Grid Rover

moves = ["N", "E", 'E', "E",  "S"]

# Format: [row, col]
# Row 0 is Top, Row 2 is Bottom. Col 0 is Left, Col 2 is Right.
pos = [0, 0]

for m in moves:
    if m == "N":
        if pos[0] > 0: # Can only go up if row > 0
            pos[0] -= 1
    elif m == "S":
        if pos[0] < 2: # Can only go down if row < 2
            pos[0] += 1
    elif m == "W":
        if pos[1] > 0: # Can only go left if col > 0
            pos[1] -= 1
    elif m == "E":
        if pos[1] < 2: # Can only go right if col < 2
            pos[1] += 1

print(f"Final Position: {pos}")
```

```python
possition = [0,0]
Movements = {'N': (-1,0),
             'S': (1,0),
             'E': (0,1),
             'W': (0,-1)}
moves = ["N", "E", "E", "E", "S"]
for move in moves:
    curren_position = [a + b for a, b in zip(possition, Movements[move])]
    
    if (0 <= curren_position[0] <= 2) and (0 <= curren_position[1] <= 2):
        possition = curren_position
        
print(possition)
```

```python
grid = [(i, j) for i in range(3) for j in range(3)]

moves = ["N", "E", "E", "E", "S", "S"]

position = [0, 0]

for move in moves:
    new_position = position.copy()
    if move == "N":
        new_position[0] -= 1
    elif move == "S":
        new_position[0] += 1
    elif move == "E":
        new_position[1] += 1
    elif move == "W":
        new_position[1] -= 1
    if tuple(new_position) in grid:
        position = new_position.copy()

print(position)
```

```python
# Solution 4: The "Diff" Checker


yesterday = [50, 40, 100, 50]
today = [50, 35, 100, 20]

changes = {}

for item_id, (old, new) in enumerate(zip(yesterday, today)):
    if old != new:
        # Store the tuple of values
        changes[item_id] = (old, new)

print(changes)
```

```python
# Solution 5: The "Best Attempt" Logger
names = ["Alice", "Bob", "Alice", "Charlie", "Bob"]
scores = [80, 75, 90, 85, 60]

final_results = {}

for name, score in zip(names, scores):
    # Logic: If name is NOT in dict, add it.
    # If name IS in dict, check if current score > stored score.
    
    if name not in final_results:
        final_results[name] = score
    else:
        if score > final_results[name]:
            final_results[name] = score

print(final_results)
```

```python
def basic_error_handling():
    # Basic try-except
    try:
        number = int(input("Enter a number: "))
        result = 100 / number
        print(f"100 divided by {number} is {result}")
    except:
        print("Something went wrong!")

    # Specific exception handling
    try:
        age = int(input("Enter your age: "))
        print(f"Next year you'll be {age + 'abc'}")
    except ValueError:
        print("That's not a valid number!")
    except ZeroDivisionError:
        print("Cannot divide by zero!")
    except Exception as e:
        print(f"An unexpected error occurred: {type(e).__name__}")


basic_error_handling()
```

```python
def complete_error_handling():
    def divide_numbers():
        try:
            numerator = float(input("Enter numerator: "))
            denominator = float(input("Enter denominator: "))
            result = numerator / denominator
        except ValueError:
            print("Please enter valid numbers!")
            return None
        except ZeroDivisionError:
            print("Cannot divide by zero!")
            return None
        else:
            # This runs only if no exception occurred
            print(f"Division successful!")
            return result
        finally:
            # This always runs
            print("Division operation completed.")
    
    result = divide_numbers()
    if result is not None:
        print(f"Result: {result}")

def file_operation_with_error_handling():
    try:
        file = open("test.txt", "r")
        content = file.read()
        print(content)
    except FileNotFoundError:
        print("File not found! Creating a new one...")
        file = open("test.txt", "w")
        file.write("This is a new file.")
        print("New file created.")
    except IOError as e:
        print(f"Input/Output error: {e}")
    except Exception as e:
        print(f"Unexpected error: {e}")
    finally:
        # Always close the file if it was opened
        if 'file' in locals() and not file.closed:
            file.close()
            print("File closed.")

# Uncomment to run
complete_error_handling()
# file_operation_with_error_handling()
```
