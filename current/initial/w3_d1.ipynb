```python
# Problem 1: The Coordinate Generator (Nested For)
# Topic: Nested for loops, append.

# Task: You are building a game board. Create a generic list called grid 
# and fill it with tuples representing coordinates (x, y).

# Iterate x from 0 to 2.

# Inside that, iterate y from 0 to 2.

# Result: [(0, 0), (0, 1), (0, 2), (1, 0), ...]

grid = []
for x in range(3):
    for y in range(3):
        grid.append((x, y))

print(grid)
```

```python
# Problem 2: The "Clipped" Normalizer

# The Scenario: You are processing sensor data stored in a matrix (list of lists).
#  The sensors have two quirks:
# They sometimes produce "noise" (negative numbers) which must be ignored completely.
# They sometimes "spike" too high. Any valid number over 50 should be capped (set) to 50.
# The Data: readings = [[10, 55, 5], [2, 60, 100], [0, -5, 25]]
# The Task: Create a single flat list called clean_data where:
# Negative numbers are removed.

# Numbers greater than 50 are replaced with 50.
# Numbers between 0 and 50 are kept as they are.
# The Data: readings = [[10, 55, 5], [2, 60, 100], [0, -5, 25]]
# Target Result: [10, 50, 5, 2, 50, 50, 0, 25]


readings = [[10, 55, 5], [2, 60, 100], [0, -5, 25]]
clean_data = []

for row in readings:
    for val in row:
        if val >= 0:  # Step 1: Filter out negatives first
            if val > 50:
                clean_data.append(50)  # Step 2: Cap values over 50
            else:
                clean_data.append(val) # Keep normal values

print(clean_data)
```

```python
# Problem 3: The Warehouse Cleaner (Nested While + Remove)
# Topic: Nested while loops, remove.
# Task: You have a list of shelves, where each shelf is a list of items: 
# warehouse = [["item", "broken", "item"], ["broken", "item"], ["item", "item"]]
# Use a while loop to go through the warehouse.
# Inside, use another while loop (or count check) to remove every instance of "broken" from the current shelf.
# Constraint: Do not create a new list; modify warehouse directly.
warehouse = [["item", "broken", "item"], ["broken", "item"], ["item", "item"]]

# Standard index-based while loop for the outer list
i = 0
while i < len(warehouse):
    shelf = warehouse[i]
    # Inner while loop to remove all instances of "broken"
    while "broken" in shelf:
        shelf.remove("broken")
    i += 1

print(warehouse)
```

```python
# Problem 4: The Priority Merger (Extend & Insert)
# Topic: for loop, extend, insert.
# Task: You have a main task list: tasks = ["Check Email", "Call Client"].
#  You have a set of new projects: 
# projects = [["Fix Bug", "Test Code"], ["Write Docs", "Push Update"]].
# Loop through the projects list.
# For the first sub-list, use insert to add its items to the beginning of tasks (index 0).
# For the subsequent sub-lists, use extend to add them to the end of tasks.
# Hint: You might need enumerate or an index tracker to know if it's the first project.




tasks = ["Check Email", "Call Client"]
projects = [["Fix Bug", "Test Code"], ["Write Docs", "Push Update"]]

for i in range(len(projects)):
    if i == 0:
        # Insert items from the first project at the start. 
        # Note: We loop through the sublist to insert them individually 
        # to keep the order, or we can use slice assignment. 
        # Let's use a nested loop with insert for clarity.
        for item in reversed(projects[i]): # Reverse so they stay in order at index 0
            print(item)
            tasks.insert(0, item)
            print(len(tasks))
            print(tasks)
    else:
        # Extend adds the whole list to the end
        tasks.extend(projects[i])

print(tasks)
```

```python
# Problem 5: The Matrix Transpose (Logic)
# Topic: Nested for loops, indexing.
# Task: Swap the rows and columns of this matrix: matrix = [[1, 2], [3, 4], [5, 6]] (3 rows, 2 columns).
# Create a new list called transposed.
# The result should look like: [[1, 3, 5], [2, 4, 6]] (2 rows, 3 columns).
# Hint: You need to loop through columns first, then rows.

matrix = [[1, 2], [3, 4], [5, 6]]
transposed = []

# We know the inner lists have length 2 (cols), so we loop range(2)
for col_index in range(2):
    print(transposed)
    new_row = []
    # Loop through the original rows to grab the item at the current column index
    for row in matrix:
        new_row.append(row[col_index])
    transposed.append(new_row)

print(transposed)
```

```python
grades = {
    "Alice": 85,
    "Bob": 72,
    "Charlie": 95,
    "Diana": 88
}

total_score = 0

# Using .items() to get both key and value
for name, score in grades.items():
    print(f"Student {name} scored {score}")
    total_score += score

average = total_score / len(grades)
print(f"Class Average: {average}")
```

```python
word = "banana"
letter_counts = {}

for letter in word:
    if letter in letter_counts:
        # Key exists, increment the value
        letter_counts[letter] += 1
    else:
        # Key is new, create it with value 1
        letter_counts[letter] = 1

# Display results using .items()
print("Character Frequencies:")
for char, count in letter_counts.items():
    print(f"{char}: {count}")
```

```python
def date_to_day_month_year(date):
    day, month, year = date.split('-')
    return day, month, year

day, month, year = date_to_day_month_year('28-01-2023')
print('Day: ', day)      # Day: 28
print('Month: ', month)  # Month: 01
print('Year: ', year)    # Year: 2023
```

```python
def build_menu(items, prices, currency="$"):
    menu = {}
    for item, price in zip(items, prices, ):
        menu[item] = f"{currency}{price}"
    return menu
```

```python

items = ["Coffee", "Tea"]
prices = [3, 2]

menu = build_menu(items, prices)
print(menu)
```

```python
db = [100, 150, 200, 300]
wh_a = [50, 80, -10, 150]
wh_b = [50, 60, 100, 140]

discrepancies = {}

for i, (target, a, b) in enumerate(zip(db, wh_a, wh_b), 1):
    # Skip data errors
    if a < 0 or b < 0:
        continue
    
    
    if a + b  != target:
        discrepancies[f"Aisle {i}"] = target - (a+b)

print(discrepancies)
```
